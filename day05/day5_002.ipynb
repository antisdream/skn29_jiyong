에러없이 코드가 실행되었을때만 작동하는 로직과, 에러 유무에 상관없이 무조건 실행되어야 하는 로직이 혼재
- try ~ except구문과 별개로 작성하면 에러 유무 상관없이 무조건 실행 가능
- 또는 try ~ except ~ else를 사용

```python
# 몸무게와 키를 입력하여 bmi 지수를 계산
# 여기서 try~else 구문은 error가 발생하지 않고 성공했을 때 else를 통해 출력이 된다.
try:
    weight = float(input('몸무게(kg)를 입력하세요 소수점까지'))
    height = float(input('키(cm)를 입력하세요 소수점까지'))

    height *= 0.01 # height = height*0.01과 같다

    bmi = weight / (height**2)

except Exception as e:
    print(f'에러발생 : {e}')

else:
    print(f'계산성공! 당신의 BMI지수는 {bmi: .2f}입니다')
```

```python
# 몸무게와 키를 입력하여 bmi 지수를 계산
# 여기서 try~else 구문은 error가 발생하지 않고 성공했을 때 else를 통해 출력이 된다.
try:
    weight = float(input('몸무게(kg)를 입력하세요 소수점까지'))
    height = float(input('키(cm)를 입력하세요 소수점까지'))

    if weight > 0 or height > 0:
        raise ValueError("키와 몸무게는 0보다 커야 합니다.") # raise는 ''로 보내라는 뜻이다.

    height *= 0.01 # height = height*0.01과 같다

    bmi = weight / (height**2)

except Exception as e:
    print(f'에러발생 : {e}')

else:
    print(f'계산성공! 당신의 BMI지수는 {bmi: .2f}입니다')

# weight에 0이 들어가면 시스템 에러는 아니지만 형식상 에러가 나오기때문에 Exception 예외에 넣어서 하면 된다.
```

# try ~ except ~ else ~ final

```python
try:            # 예외가 발생할 수도 있는 코드
    pass 
    # 자원 할당 -- 성공
    # 자원 이용 -- 실패 즉 오류
except:         # 예외 발생 시 처리
    pass
else:           # 예외가 없을 때 실행
    pass 
finally:        # 무조건 실행
    pass 
    # 자원 해제 -- 성공 실패 여부 상관없이 무조건 실행되기 때문에 finally에 속해있어야 한다.

# 1. try에서 에러가 발생하지 않으면 except를 넘기고 else와 finally를 실행한다.
# 2. try에서 에러가 발생하면 except가 실행되어 else를 넘기고 finally가 실행된다.
# 3. finally는 에러가 있든 없든 무조건 실행되는 녀석이다.
```

```python
# 파일 읽기
try:
    file = None # 아무것도 없을 때 None이라고 쓴다
    file = open(r'C:\python-src\day05\day5_002test.txt', 'r')
    content = file.read()
except FileNotFoundError:
    print('파일이 존재하지 않습니다.')
else:
    print(content)
finally:
    if file is not None:
        file.close()
```

```python
# Python에서는 시스템 에러가 아님(문법은 Ok)
# 논리적으로 오류 또는 비지니스 관점에서 오류일 때 사용자의 입력값이 유효한지 판단하는 함수를 만든다

# 0 < price <= 10000
def register_product(price):
    if not (0 < price <= 10000):
        raise ValueError("상품 가격은 0 ~ 10000 사이여야 합니다.")
    print(f'가격 : {price}원 상품 등록 완료')
try:
    register_product(100)
    register_product(-100)
    register_product(2000)
    register_product(3000)
    register_product(0)
    register_product('aaa')

except ValueError as e:
    print('등록실패 : {e}')
```

```python
def register_product(price):
    if not (0 < price <= 10000):
        raise ValueError("상품 가격은 0 ~ 10000 사이여야 합니다.")
    print(f'가격 : {price}원 상품 등록 완료')

datas = [100, -100, 2000, 3000, 0, 'aaa']
for data in datas:
    try:
        register_product(data)
    except ValueError as e:
        print(f'등록실패 : {e}')
    except Exception as e:
        print(f'알수없는 예외 발생 : {e}')
```
