```python
# 옛날에는...
print('---------- 나이 입력 ----------')
input_number = int(input('숫자를 입력하세요'))

for i in input_number:
    print(type(i) == str)

print(f' 입력한 숫자는 : {input_number}')
```

```python
# 정확히 정수나 문자를 혼용해서 사용하면 문제 없지만
# 소수점이 포함된 실수를 사용하면 걸러내지 못한다..

#사용자에게 숫자를 입력받아서 +1을 하는 로직
print('---------- 나이 입력 ----------')
input_number = input('숫자를 입력하세요')

is_number = True

#for문의 역할은 숫자가 아닌녀석을 찾으려고 할때 쓰인다.
for i in input_number: # 100, 100r, 1t200, 12.5 등 입력을 해보며 에러를 찾는다
    if not i.isdecimal():
        is_number = False
        break

if is_number and not input_number.count('.'): 
    input_number = int(input_number)
    print(f' 입력한 숫자에 +1은 : {input_number + 1}')
else:
    print('정수만 입력하세요')
```

```python
print('---------- 나이 입력 ----------')

try:
    input_number = int(input('숫자를 입력하세요'))
    print('잘했어요')
# try 구문은 완벽하게 실행이 되면 밑의 except로 넘어가지 않고 끝낸다.
# 입력이 잘못됐을 때 except로 넘어가버린다.
except:
    print('정수만 입력되어야 합니다.')

#아무거나 입력해도 죽지 않는 무적
```

```python
print('---------- 나이 입력 ----------')

try:
    temp = [1,2,3]
    print(temp[2])
    input_number = int(input('숫자를 입력하세요'))
    print('잘했어요')
except:
    print('정수만 입력되어야 합니다.')
```

```python
print('----------Value와 Index Error묶음-----------')

try:
    temp = [1,2,3]
    print(temp[1]) # indexError
    input_number = int(input('숫자를 입력하세요')) # valueError
    print('잘했어요')
except ValueError:
    print('정수만 입력되어야 합니다.')
except IndexError:
    print('인덱스 범위를 체크하세요.')
```
