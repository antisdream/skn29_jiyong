```python
class Dog:
    def __init__(self, name, bread, age, is_health):
        self.name = name
        self.bread = bread
        self.age = age
        self.is_health = is_health
    def showInfo(self):
        print(f'name:{self.name}, bread:{self.bread}, age:{self.age}, is_health:{self.is_health}')


# 객체생성
d1 = Dog('누렁이', '불독', 2, True)
d1.showInfo()
```

```python
d2 = Dog('꾸꾸', '아메리칸블리', 3, True)
d2.showInfo()
```

```python
# 학생클래스 : Student
# 속성(변수) : 이름, 나이, 주소, 국,영,수
# 메소드 : 평균, 총점, 정보출력(모든변수 다 출력)
class Student:
    def __init__(self,age,name,addr,kor,eng,math): 
        self.name = name
        self.age = age
        self.addr = addr
        self.kor = kor
        self.eng = eng
        self.math = math

    def getAverage(self):
        return self.getTotal() / 3
# getTotal로 묶는 이유가 나중에 총점을 구하는 수식이 바뀔 경우 하나만 고쳐주면 된다.
# 함수 순서상 getTotal이 getAverage보다 앞에 나와야하는데 class에서는 상관없다
    def getTotal(self):
        return self.kor + self.eng + self.math
```

```python
s1 = Student(18, '홍길동', '서울시 구로구', 98, 97, 85)
# 평균
print(s1.getAverage())
# 총점
print(s1.getTotal())
```

```python
# 생성자(def ~ self) 밖에서 만드는 함수는 공용 함수...
# 쇼핑몰... 쇼핑카트(장바구니)
# 클래스 이름 ShoppingCart
# 변수 items라는 리스트, 쇼핑카트주인의 이름
# 메소드 add_item

class ShoppingCart:
    def __init__(self,owner):
        self.owner = owner
        self.items = [] #안에 물건을 함수 밖에서 가져왔기 때문에 생성자 ()에는 안들어감
    def add_item(self,item): # class안에 있기 때문에 첫번째엔 무조건 self
        #pass 추가할 물건이 생각 안나면 pass로 넘기기
        self.items.append(item) #list에서 추가하는 함수 .append()

# 철수와 영희가 각각 빈 장바구니를 하나씩 집어 들었다
cs = ShoppingCart("철수")
yh = ShoppingCart("영희")

# 철수가 쇼핑을 시작
cs.add_item('노트북')
cs.add_item('마우스')

# 철수의 쇼핑목록
print(f'{cs.owner}의 장바구니 : {cs.items}')

# 영희의 쇼핑목록
print(f'{yh.owner}의 장바구니 : {yh.items}')
```
