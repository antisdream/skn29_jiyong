```python
#이런 코드가 있다.

def choice_menu():
    print('메뉴를 선택하세요')
    print("1. 입력 3. 조회")
    print('2. 출력 4. 종료')
    choice = int(input('번호를 선택하세요'))
    return choice

while True:
    choice = choice_menu()
    if choice == 1:
        print('정보를 입력하세요')
    elif choice == 2:
        print(" 사용자 정보 시스템 입니다.")
    elif choice == 3:
        print(" 조회할 번호를 입력하세요")
    elif choice == 4:
        print(" 프로그램을 종료합니다.")
        break
```

```python
class fishbread:
    def __init__(self, filling):
        # 팥/크림 같은 속성
        self.filling = filling
        self.baked = True

    def pour_batter(self):
        # 반죽 붓기 동작
        print("반죽을 붓습니다.")

    def add_filling(self):
        # 속 넣기 동작
        print(f"{self.filling}를 넣습니다.")

    def bake(self):
        # 굽기 동작 (method)
        self.baked = True
        print("붕어빵을 굽습니다. 완성!")

# 만드는 사람(생성자 호출) -> 인스턴스 생성
a = fishbread("팥")
b = fishbread("크림")
```

```python
a.pour_batter()
a.add_filling()
a.bake()
```

```python
b.pour_batter()
b.add_filling()
b.bake()
```
