# [1교시]

#  HTML5 문서 구조와 시맨틱 마크업

## 1. 학습 목표
* 웹 브라우저가 HTML 문서를 요청하고 화면에 그리기까지의 기본적인 파싱 및 렌더링 과정을 이해합니다.
* HTML 문서의 표준 구조와 필수적인 메타 데이터 태그들의 의미와 역할을 설명할 수 있습니다.
* 무의미하게 요소를 감싸는 `div` 위주의 구조화와 의미를 내포한 시맨틱 태그(Semantic Tags) 구조화의 차이를 비교하고, 이에 따른 검색 엔진 최적화(SEO) 및 웹 접근성 향상 효과를 이해합니다.
* `h1`부터 `h6`까지의 제목 태그 규칙 및 문서 개요(Outline) 생성 원칙을 준수하여 뼈대 설계를 할 수 있습니다.

---

## 2. 브라우저 렌더링 맛보기 (요청부터 화면 표시까지)
웹 사용자가 브라우저 주소창에 URL을 입력하면 다음과 같은 과정을 거쳐 화면에 결과물이 렌더링됩니다.

```
[사용자 요청] -> (서버로 HTTP Request 전송)
     ↓
[서버 응답]   -> (HTML, CSS, JS 파일 등 HTTP Response 반환)
     ↓
[구문 분석]   -> 1. HTML 파싱 ──> DOM 트리 생성
(Parsing)        2. CSS 파싱  ──> CSSOM 트리 생성
     ↓
[렌더 트리]   -> DOM + CSSOM 결합 (화면에 보이지 않는 display: none 요소 등은 제외)
(Render Tree)
     ↓
[레이아웃]    -> 각 노드의 정확한 위치와 크기 계산 (Reflow)
(Layout)
     ↓
[페인팅]      -> 계산된 값을 바탕으로 화면에 실제 픽셀로 변환 (Repaint)
(Paint)
```

* **HTML 파싱**: 브라우저가 위에서 아래로 한 줄씩 HTML 코드를 읽으면서 기계가 이해할 수 있는 트리 구조인 **DOM(Document Object Model) 트리**를 구축합니다.
* **렌더링**: 완성된 DOM 트리와 CSSOM(CSS Object Model) 트리가 결합하여 사용자 화면에 실제로 그려지게 됩니다. 잘못 작성된 마크업(예: 닫는 태그 누락 등)은 브라우저가 임의로 해석하여 오류를 복구하려 하므로 비효율적이고 비정상적인 출력을 야기할 수 있습니다.

## 3. HTML5 문서의 표준 구조 (뼈대 파헤치기)
가장 표준적인 HTML5 문서 구조는 아래와 같이 선언 및 구성됩니다.

```html
<!DOCTYPE html>
<html lang="ko">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>문서 제목</title>
</head>
<body>
  <!-- 화면에 나타나는 실제 콘텐츠가 들어갑니다 -->
</body>
</html>
```

### 1) `<!DOCTYPE html>`
* **역할**: 현재 문서가 표준 **HTML5** 문서임을 브라우저에 알리는 선언문입니다.
* **중요성**: 생략되거나 잘못 선언될 경우, 브라우저는 하위 호환성을 유지하기 위해 **쿼크 모드(Quirks Mode)**로 렌더링을 시도하며, 이로 인해 최신 CSS/레이아웃이 정상적으로 작동하지 않을 수 있습니다.

### 2) `<html lang="ko">`
* **역할**: 웹 문서의 주 언어가 한국어(`ko`)임을 나타냅니다.
* **장점**: 화면 낭독기(스크린 리더)를 사용하는 시각 장애인 사용자가 접속했을 때 알맞은 언어 발음(엔진)으로 읽어 주도록 도우며, 검색 엔진이 해당 국가나 언어 기반 검색 시 적절하게 색인화하도록 기여합니다.

### 3) `<head>` 영역
* 브라우저 화면에 직접 렌더링되지는 않지만, 문서의 메타데이터(정보), 캐릭터셋 인코딩 설정, 뷰포트 설정, 외부 리소스(CSS, JS 등) 링크 정보 등을 포함합니다.
* **`<meta charset="UTF-8">`**: 문자 인코딩 방식으로 전 세계 모든 문자를 표현할 수 있는 UTF-8을 지정하여 한글 깨짐 현상을 방지합니다.
* **`<meta name="viewport" content="width=device-width, initial-scale=1.0">`**: 반응형 웹 디자인을 지원하기 위해 필수적입니다. 기기의 가로 해상도 크기(`width=device-width`)에 맞게 웹 페이지의 너비를 설정하고, 초기 확대/축소 배율을 1배(`initial-scale=1.0`)로 고정하여 화면이 모바일이나 태블릿에서 비정상적으로 작게 표현되는 것을 방지합니다.
* **`<title>`**: 브라우저 탭에 표시되는 제목이자 검색 엔진 결과 페이지에서 가장 돋보이는 제목 텍스트로 활용됩니다.

### 4) `<body>` 영역
* 실제 웹 브라우저 뷰포트에 표시될 텍스트, 이미지, 링크, 동영상 등의 모든 visual 요소가 작성되는 본문 영역입니다.

## 4. 시맨틱 마크업(Semantic Markup)이 필요한 이유
'시맨틱(Semantic)'이란 **"의미가 통하는"** 또는 **"의미론적인"**이라는 뜻을 갖습니다. 구조화되지 않은 일반 상자인 `<div>`만으로 도배된 레이아웃 대신, 각각의 역할에 맞는 이름을 가진 태그들을 사용하는 마크업 방식을 의미합니다.

###  왜 시맨틱 마크업을 사용해야 할까요?
1. **검색 엔진 최적화 (SEO)**: 검색 엔진 로봇이 웹 페이지를 크롤링할 때 어떤 부분이 헤더이고 본문인지, 가장 중요한 주요 콘텐츠 영역이 어디인지 정확하게 분류하여 검색 노출 우선순위 결정에 반영합니다.
2. **웹 접근성(Web Accessibility)**: 시각 장애인이 화면 낭독기(스크린 리더)를 통해 웹 서핑을 할 때, 시맨틱 태그는 중요한 탐색 지점(Landmark)을 제공합니다. 사용자는 메뉴(`<nav>`)나 핵심 본문(`<main>`)으로 바로 건너뛰어 청취할 수 있습니다.
3. **개발자 협업 및 유지보수 가독성**: 코드만 보더라도 특정 영역이 전체 페이지의 꼬리말인지, 사이드바인지 직관적으로 이해할 수 있어 유지보수 비용이 대폭 감소합니다.

## 5. 핵심 HTML5 시맨틱 태그 정리

| 태그 | 의미 및 역할 | 주의사항 및 특징 |
|:---|:---|:---|
| `<header>` | 문서나 특정 섹션의 머리말 영역 (주로 로고, 검색창, 내비게이션 포함) | 문서 전체뿐 아니라 `<article>` 등 개별 섹션 내에도 포함될 수 있음 |
| `<nav>` | 다른 페이지나 문서 내부의 다른 영역으로 연결되는 링크 링크 집합 (내비게이션 바, 목차) | 주요 링크에만 사용 (푸터의 단순 약관 링크 등에는 생략 권장) |
| `<main>` | 문서의 핵심 및 고유 콘텐츠 영역 | 문서 내 **단 하나만** 존재해야 함 (IE 미지원 대비로 필요 시 role="main" 명시하기도 함) |
| `<section>` | 문서에서 독립적인 장(Chapter)이나 테마별 콘텐츠 단락을 묶어주는 영역 | 일반적으로 내부에 반드시 제목 태그(`<h2>`~`<h6>`)를 포함하는 것이 권장됨 |
| `<article>` | 뉴스 기사, 블로그 포스트, 포럼 글처럼 그 자체로 독립적인 배포/재사용이 가능한 콘텐츠 | 독립적으로 다른 사이트에 퍼가더라도 문맥상 흐름이 자연스러운 완전한 정보 단락에 사용 |
| `<aside>` | 주요 본문 흐름 외의 주변부 정보 콘텐츠 (광고, 카테고리 목록, 스크롤 탑 버튼 등) | 본문 흐름을 방해하지 않는 주변 콘텐츠에 사용 |
| `<footer>` | 문서 또는 특정 섹션의 바닥글 영역 (저작권 표시, 제작 정보, 패밀리 사이트 등) | 주소나 연락처는 푸터 내에 `<address>` 태그와 함께 사용하는 경우가 많음 |

## 6. 제목 태그(`<h1>` ~ `<h6>`)의 계층 구조 규칙
제목 태그는 브라우저가 웹 문서의 계층(Outline) 구조를 해석할 때 척도가 됩니다.

* **원칙 1: 한 페이지당 `<h1>`은 가급적 하나만 권장됩니다.**
  * 전체 페이지의 대제목이나 로고 영역에 주로 배치됩니다.
* **원칙 2: 계층 구조 순서를 건너뛰어서는 안 됩니다.**
  * `<h1>` 다음에 바로 `<h3>`이 와서는 안 되며, 반드시 부모-자식 흐름에 맞게 `<h2>`를 거쳐 `<h3>`으로 내려가야 합니다. (올바른 순서: `h1` -> `h2` -> `h3` -> `h2` ...)
* **원칙 3: 크기를 줄이기 위한 목적으로 제목 태그 계층을 낮춰 잡아서는 안 됩니다.**
  * 디자인적인 폰트 크기 변경은 CSS의 `font-size` 속성을 통해 처리해야 합니다.

---

## 7. `div` / `span` vs 시맨틱 태그 사용 기준

* **`div` (Block 요소)**: 의미 없이 구역을 묶을 때만 사용합니다. CSS 레이아웃 스타일링(그루핑, 간격 주기)이나 JS 조작을 위한 타깃 컨테이너로 삼을 때 적합합니다.
* **`span` (Inline 요소)**: 텍스트 줄바꿈 없이 본문 내부의 특정 구절, 단어에만 개별 스타일을 입히거나 스크립트 처리를 하고자 할 때 감싸는 요소입니다.
* **시맨틱 태그**: 의미가 존재하는 단락이나 콘텐츠 영역은 **무조건** 해당 시맨틱 태그로 작성하고, 이를 감싸거나 스타일적으로 보완하는 껍데기 역할에만 `div`를 사용합니다.

---

## 8. 핵심 키워드 정리
* **DOCTYPE**: 브라우저에게 HTML5 문서 규격 준수를 명시하는 선언문.
* **viewport**: 모바일/태블릿 등의 화면 비율과 배율을 제어하는 메타 정보.
* **시맨틱 태그**: 의미론적 태그로 기계와 인간 모두에게 콘텐츠 구조를 명확히 이해시킴.
* **문서 개요 (Outline)**: 제목 태그(`h1~h6`) 및 섹셔닝 태그를 기준으로 형성되는 문서의 계층적 구조.

# [2교시]

# HTML5 핵심 태그 (텍스트·표·폼·미디어)

## 1. 학습 목표
* 텍스트의 성격과 용도에 부합하는 웹 표준 텍스트 태그를 선택하여 구분해 쓸 수 있습니다.
* 정보를 행과 열로 구조화하여 표시하는 `table` 태그의 명확한 골격을 잡고 셀 병합(`colspan`, `rowspan`)을 사용할 수 있습니다.
* 하이퍼링크(`<a href>`)와 이미지(`<img src>`) 삽입 시 접근성을 위한 필수 속성(`alt`)을 올바르게 기입합니다.
* 사용자의 입력을 받아 백엔드 서버로 전송하는 `form` 태그 및 `input`의 핵심 타입들을 이해하고, `label`의 `for` 연동을 통한 웹 접근성 표준 규격을 만족합니다.
* 미디어 영역의 구조적 정의를 돕는 `<figure>`, `<figcaption>`과 `<video>`, `<audio>` 등의 미디어 삽입 원리를 익힙니다.

---

## 2. 텍스트 태그 완전 정리
HTML은 단순히 텍스트 크기를 키우기만 하는 것이 아니라 문맥(Context)에 따라 의미가 강조되는 별도의 태그를 지원합니다.

| 태그 | 역할 | 사용 시 특징 |
|:---|:---|:---|
| `<p>` | 하나의 독립된 문단(Paragraph)을 나눕니다. | 상/하단에 기본 마진(Margin)이 자동으로 설정되는 Block 요소입니다. |
| `<span>` | 문장 내 특정 단어나 글자를 묶어 디자인을 적용합니다. | 줄바꿈이 없는 Inline 요소로 단순 그루핑 목적에 가장 적합합니다. |
| `<strong>` | **텍스트의 내용상 아주 중요함**을 의미론적으로 강조합니다. | 화면 낭독기에서도 음성의 톤이나 크기를 조절해 중요성을 전달합니다. |
| `<em>` | **텍스트를 구어적으로 강조**할 때 사용합니다. (이탤릭체 표시) | Emphasis의 약자로 단순 폰트 기울임이 아닌 구어적 강세를 부여합니다. |
| `<b>` / `<i>` | 시각적으로만 굵게(Bold) 하거나 기울이게(Italic) 합니다. | **의미적 강조가 전혀 없습니다.** 최신 표준에서는 `strong`, `em` 권장. |
| `<br>` | 줄바꿈(Line Break)을 수행합니다. | 줄 간격을 띄우는 용도로 남발하면 안 되며 문단 구분은 `<p>`를 써야 합니다. |
| `<hr>` | 이야기 주제가 바뀔 때 시각적인 구분선을 긋습니다. | Horizontal Rule의 약자로, 현대 웹 디자인에서는 구분선 뿐 아니라 주제 전환점 의미도 갖습니다. |

## 3. 목록(List) 태그 정리
목록을 그루핑하면 구조화가 명확해지며 CSS 메뉴 바나 게시판 리스트의 원형이 됩니다.

```html
<!-- 1. 비순서형 목록 (Unordered List) -->
<ul>
  <li>사과</li>
  <li>바나나</li>
</ul>

<!-- 2. 순서형 목록 (Ordered List) -->
<ol>
  <li>첫 번째 단계</li>
  <li>두 번째 단계</li>
</ol>

<!-- 3. 정의형 목록 (Description List) -->
<dl>
  <dt>용어</dt>
  <dd>용어에 대한 설명이 들어갑니다.</dd>
</dl>
```
* `<dt>`는 Definition Term(용어의 이름), `<dd>`는 Definition Description(용어의 상세 설명)을 의미합니다. 사전식 설명이나 정보-값 매칭 시 유용합니다.

---

## 4. 표(Table) 태그 구조의 시각화
표는 구조적 해석이 어렵기 때문에 스크린 리더 등 웹 접근성을 고려하여 `thead`, `tbody`, `tfoot` 등으로 명확히 구획하여 마크업해야 합니다.

### 1) 구조도
```
+------------------------------------------+  <- <table>
|                 제목 행                   |  <- <thead>  (th: 굵은 제목 셀)
+------------------------------------------+
|                 내용 행                   |  <- <tbody>  (td: 일반 데이터 셀)
|                 내용 행                   |
+------------------------------------------+
|                 합계 행                   |  <- <tfoot>
+------------------------------------------+
```

### 2) 셀 병합 속성
* **`colspan="N"`**: 열(Column)을 병합합니다. 가로로 인접한 N개의 셀을 하나로 합칩니다.
* **`rowspan="N"`**: 행(Row)을 병합합니다. 세로로 인접한 N개의 셀을 하나로 합칩니다.

## 5. 링크와 이미지
### 1) 하이퍼링크 `<a>`
* `href` 속성에 이동하고자 하는 목적지 URL을 정의합니다.
* `target="_blank"`를 사용하면 현재 탭이 아닌 새 탭에서 링크를 띄웁니다.
  * *보안 팁*: 새 창 이동 시 `rel="noopener noreferrer"` 속성을 함께 기입하여 부모 창 객체 조작 취약점을 방지합니다.

### 2) 이미지 `<img>`
* `<img src="이미지경로" alt="대체텍스트">`
* **`alt` 속성의 필수성**: 이미지가 로드 실패 시 깨진 그림 대신 대체 텍스트를 제공합니다. 또한 검색 엔진 봇에게 해당 이미지가 어떤 이미지인지를 이해시키며, 시각장애인용 화면 낭독기에서 이미지 대신 읽어 주어 필수적인 대체 정보 역할을 수행합니다. `alt`를 아예 지우는 것은 웹 접근성 위반입니다.

## 6. 폼(Form) 태그와 입력 양식
사용자 정보 입력, 파일 전송, 가입 폼 등은 백엔드와의 통신을 위한 가장 핵심적인 관문입니다.

### 1) `<form>` 속성
* `action`: 입력받은 양식의 데이터를 최종적으로 처리할 백엔드 서버의 URL 주소.
* `method`: 전송 방식 (`GET`: URL 파라미터에 값을 노출하여 전송, `POST`: HTTP Body 패킷 영역에 숨겨서 대용량/보안 전송).

### 2) `<label>`과 `<input>`의 연결
웹 접근성을 높이기 위해 입력창 옆의 글자(Label)를 누르더라도 자동으로 인풋창이 활성화(포커스)되도록 설정해야 합니다.

```html
<!-- label의 for와 input의 id를 일치시킵니다 -->
<label for="user-email">이메일 주소</label>
<input type="email" id="user-email" name="email">
```

### 3) 주요 `<input>` type 속성
* `text`: 일반 한 줄 텍스트 입력
* `password`: 입력 문자가 가려짐(`*` 또는 `●`)
* `email`: 메일 형식 검증 기능 자동 수행
* `tel`: 모바일 기기 접속 시 숫자 키패드 팝업 유도
* `radio`: 그룹 내 단 하나만 선택 가능 (`name` 속성값이 같아야 그룹으로 묶임)
* `checkbox`: 여러 개 다중 선택 가능
* `date`: 달력 형식의 날짜 선택 창 표출
* `submit`: 폼 데이터를 서버로 송신하는 전송 버튼

# CSS3 적용 방법과 선택자

## 1. 학습 목표
* CSS를 웹 페이지에 적용하는 3가지 방식의 차이점을 파악하고, 실무에서 외부 스타일시트(`.css`) 분리 적용 방식을 지향해야 하는 이유를 이해합니다.
* 전체 선택자, 태그 선택자, 클래스 및 ID 선택자의 정확한 구문 규칙을 습득합니다.
* 자손, 자식, 인접 형제, 일반 형제 선택자 등 결합 선택자(Combinators)를 조합하여 원하는 대상 요소를 입체적으로 조작합니다.
* 속성 선택자(`[attr=value]`) 및 다양한 마우스 반응/구조 가상 클래스(`:hover`, `:nth-child()`), 그리고 가상 요소(`::before`, `::after`)를 응용합니다.
* CSS 캐스케이딩(Cascading)의 핵심 원리인 우선순위 점수(Specificity)를 수학적으로 계산할 수 있고, 유지보수를 방해하는 `!important` 구문을 피해야 하는 사유를 파악합니다.

---

## 2. CSS를 적용하는 3가지 방법 비교

| 방식 | 작성 형태 | 장점 | 단점 | 실무 사용 권장도 |
|:---|:---|:---|:---|:---|
| **인라인 스타일 (Inline Style)** | `<div style="color: red;">` | 특정 단일 요소에 스타일을 가장 빠르게 임시로 적용 가능 | HTML과 CSS의 관심사 분리가 무너져 가독성이 저하되며 유지보수가 극도로 불가능함 | **매우 비권장** (JS 동적 변경 제외) |
| **내부 스타일시트 (Embedded Style)** | `<head>` 안에 `<style>` 태그 작성 | 별도의 파일 없이 해당 단일 HTML 문서의 모든 스타일 한 번에 관리 | 다수의 HTML 페이지가 존재할 때 중복 코드가 대량 발생하여 스타일 통제가 불가능함 | **소규모 샘플/단일 테스트** 시에만 제한 권장 |
| **외부 스타일시트 (External Link)** | `<link rel="stylesheet" href="style.css">` | 여러 HTML 문서에서 **동일한 단일 CSS 파일을 캐싱 및 공유**하므로 용량이 절감되고 전체 디자인을 한 번에 제어 가능 | 서버 요청 횟수가 늘어날 수 있음 (현대 브라우저 환경에서는 무시할 만함) | **실무 표준 (적극 권장)** |

---

## 3. 기본 선택자 (Basic Selectors)

* **전체 선택자 (`*`)**: 페이지 내의 모든 요소를 선택합니다. 주로 여백 초기화(CSS Reset) 시 적용됩니다.
* **태그 선택자 (`태그명`)**: 해당 태그명을 가진 모든 HTML 요소를 일괄 선택합니다. (예: `p { color: #333; }`)
* **클래스 선택자 (`.클래스명`)**: HTML에서 `class="..."` 속성을 부여한 요소를 복수 타깃으로 선택합니다. (예: `.btn { border-radius: 4px; }`)
* **ID 선택자 (`#ID명`)**: HTML 문서 내에서 고유해야 하는 `id="..."` 요소를 단일 타깃으로 선택합니다. (예: `#main-header { height: 80px; }`)

---

## 4. 결합 선택자 (Combinators)
각 요소의 포함 관계 및 순서적 형제 관계를 기반으로 대상을 정밀 타격합니다.

* **자손 선택자 (`A B`)**: A 요소 **내부에 중첩된 모든 하위** B 요소를 선택합니다. (단계 깊이에 상관없음)
* **자식 선택자 (`A > B`)**: A 요소 **바로 밑에 위치한 직속 자식**인 B 요소만 선택합니다. (중첩 단계 1단계만 해당)
* **인접 형제 선택자 (`A + B`)**: A 요소의 **바로 뒤에 인접해서 달라붙은 첫 번째 형제** B 요소 하나만 선택합니다.
* **일반 형제 선택자 (`A ~ B`)**: A 요소의 **뒤에 위치한 모든 형제** B 요소를 전부 선택합니다.

---

## 5. 속성 선택자 (Attribute Selectors)
클래스나 ID 외에 태그 내부의 속성값 조건에 대응하여 선택합니다.

* **`[href]`**: `href` 속성이 존재하는 모든 요소를 선택합니다.
* **`[type="text"]`**: `type` 속성값이 정확히 `"text"`인 요소만 선택합니다.
* **`[href^="https"]`**: `href` 속성이 `"https"`로 **시작하는** 요소를 선택합니다.
* **`[href$=".pdf"]`**: `href` 속성이 `".pdf"`로 **끝나는** 요소를 선택합니다.
* **`[class*="btn-"]`**: 클래스명에 `"btn-"`이라는 문자열이 **포함된** 모든 요소를 선택합니다.

---

## 6. 가상 클래스(Pseudo-Classes)와 가상 요소(Pseudo-Elements)

### 1) 주요 가상 클래스 (상태 및 구조 기반)
* **`:hover`**: 요소 위에 마우스 포인터가 올라갔을 때 적용합니다.
* **`:focus`**: 폼 인풋 양식 등이 키보드 탭이나 마우스 클릭으로 입력 활성화 상태가 되었을 때 적용합니다.
* **`:first-child`**: 부모의 첫 번째 자식 요소를 선택합니다.
* **`:last-child`**: 부모의 마지막 자식 요소를 선택합니다.
* **`:nth-child(N)`**: 부모의 N번째 자식 요소를 선택합니다.
  * `:nth-child(even)` (짝수 번째), `:nth-child(odd)` (홀수 번째), `:nth-child(3n)` (3의 배수 번째) 등 수식을 넣을 수 있어 표 스타일링 등에 자주 쓰입니다.

### 2) 주요 가상 요소 (문서 트리 외 가상 콘텐츠 생성)
* **`::before`**: 선택한 요소의 **가장 앞부분**에 가상의 자식 콘텐츠를 동적으로 추가합니다.
* **`::after`**: 선택한 요소의 **가장 뒷부분**에 가상의 자식 콘텐츠를 동적으로 추가합니다.
  * **필수 사항**: `::before`와 `::after` 사용 시에는 CSS 안에 빈 문자열이라도 좋으니 **`content: "";`** 속성을 반드시 선언해야 렌더링에 노출됩니다. 주로 아이콘, 뷸렛 기호, 데코레이션 선을 넣을 때 사용합니다.

---

## 7. 우선순위(Specificity) 점수 계산법
여러 CSS 규칙이 동일한 요소를 다르게 스타일링할 때, 브라우저는 우선순위 점수를 계산하여 가장 점수가 높은 스타일을 적용합니다.

### 1) 점수 등급표
1. **`!important`**: 99999점 (우선순위 체계를 교란하므로 실무 사용 극도 기피 대상)
2. **인라인 스타일 (Inline Style)**: 1000점 (HTML 내부에 적은 스타일)
3. **ID 선택자 (`#`)**: 100점
4. **클래스 선택자 (`.`), 속성 선택자, 가상 클래스**: 10점
5. **태그 선택자, 가상 요소**: 1점
6. **전체 선택자 (`*`)**: 0점

### 2) 계산 예시
* `div p` -> 태그 2개 = **2점**
* `div .card p` -> 태그 2개(2점) + 클래스 1개(10점) = **12점**
* `#main-nav .menu-item a` -> ID 1개(100점) + 클래스 1개(10점) + 태그 1개(1점) = **111점**
* `#main-nav .menu-item a:hover` -> ID 1개(100점) + 클래스 1개(10점) + 가상 클래스 1개(10점) + 태그 1개(1점) = **121점**

> [!WARNING]
> 점수가 같을 경우, **가장 마지막(아래쪽)에 선언된 스타일**이 캐스케이딩(Cascading) 원칙에 따라 앞서 기술된 스타일을 덮어씁니다.

---

## 8. 핵심 키워드 정리
* **specificity**: 선택자가 가지는 고유의 가중치 점수.
* **가상 클래스 (Pseudo-Classes)**: 동작 상태나 순서 조건에 반응하는 일종의 상태 필터.
* **가상 요소 (Pseudo-Elements)**: 마크업상 실제로 존재하지 않는 가짜 요소를 생성하여 추가하는 기법.
* **결합 선택자**: 포함 및 인접 관계를 나타내어 CSS 가독성과 정밀성을 향상시키는 방법.

# 박스 모델, 텍스트·배경 스타일링

## 1. 학습 목표
* CSS 박스 모델(Box Model)의 4대 요소인 콘텐츠, 패딩, 테두리, 마진의 개념을 명확히 정의합니다.
* `box-sizing` 속성의 `content-box`와 `border-box` 차이를 픽셀 크기 계산 수식으로 증명하고, 현업에서 `border-box`를 기본 리셋으로 설정하는 명확한 당위성을 파악합니다.
* `display` 속성에 따른 블록(Block), 인라인(Inline), 인라인 블록(Inline-block), 그리고 감추기(None)의 물리적인 렌더링 성격을 비교 이해합니다.
* px(절대 단위), %(상대 단위), em/rem(폰트 비례 단위), vw/vh(뷰포트 비례 단위) 등 CSS 핵심 단위를 용도에 맞게 적용합니다.
* 타이포그래피(글꼴 스택, 굵기, 줄간격 등) 및 배경 제어(background-image, size, position), 박스 입체감 처리(border-radius, box-shadow) 기법을 습득합니다.

---

## 2. CSS 박스 모델 (Box Model) 구조
브라우저는 모든 HTML 요소를 사각형 모양의 상자(Box)로 규정하여 렌더링 영역을 계산합니다.

```
+------------------------------------------+  <- 바깥 여백 (Margin)
|  +------------------------------------+  |  <- 테두리 선 (Border)
|  |  +------------------------------+  |  |  <- 안쪽 여백 (Padding)
|  |  |                              |  |  |
|  |  |       콘텐츠 (Content)       |  |  |  <- 실제 텍스트나 이미지 영역
|  |  |                              |  |  |
|  |  +------------------------------+  |  |
|  +------------------------------------+  |
+------------------------------------------+
```

* **콘텐츠 (Content)**: 글자나 이미지 등 요소의 실제 내용이 위치하는 공간입니다. (`width`, `height`로 제어)
* **패딩 (Padding)**: 콘텐츠와 테두리 선(Border) 사이의 안쪽 여백입니다. 패딩을 주면 박스가 내부적으로 팽창하며 요소의 배경색이 패딩 영역까지 칠해집니다.
* **테두리 (Border)**: 안쪽 여백과 바깥 여백을 가르는 경계선입니다. 선의 두께, 스타일, 색상을 가집니다.
* **마진 (Margin)**: 다른 인접 요소들과의 간격을 벌리기 위한 테두리 밖의 순수 바깥 여백입니다. 다른 요소와 배경 영역에 영향을 주지 않는 빈 공간입니다.

---

## 3. box-sizing 속성비교 (박스 크기 계산법)

어떤 요소를 가로 300px, 패딩 20px, 테두리 5px로 설정했을 때 브라우저가 차지하는 실제 가로 폭은 `box-sizing` 설정에 따라 완전히 달라집니다.

### 1) `content-box` (기본값)
* **정의**: `width` 속성으로 지정한 크기가 오직 **콘텐츠 영역**에만 한정됩니다.
* **실제 계산 가로 크기**: `width` + `left padding` + `right padding` + `left border` + `right border`
  * 예시 계산: $300\text{px} + 20\text{px}\times 2 + 5\text{px}\times 2 = 350\text{px}$
* **단점**: 패딩이나 테두리 두께를 수정할 때마다 전체 레이아웃 가로 너비가 비정상적으로 깨져 매번 계산을 다시 해야 합니다.

### 2) `border-box` (실무 표준 권장)
* **정의**: `width` 속성으로 지정한 크기 안에 **콘텐츠, 패딩, 테두리가 모두 포함**됩니다.
* **실제 계산 가로 크기**: 무조건 `width`로 고정됩니다. (콘텐츠 영역 크기가 패딩/테두리 공간만큼 알아서 축소됨)
  * 예시 계산: 콘텐츠 너비($250\text{px}$) + 패딩($40\text{px}$) + 테두리($10\text{px}$) = 딱 $300\text{px}$
* **장점**: 디자이너가 요청한 가이드라인 크기 그대로 수치를 지정하고 안쪽 여백을 자유롭게 조절해도 레이아웃의 전체 구조가 전혀 흐트러지지 않습니다.

---

## 4. display 속성: block vs inline vs inline-block

| 속성값 | 개행 여부 | 가로 너비 (width) | 마진/패딩 영향 | 주요 태그 |
|:---|:---|:---|:---|:---|
| **`block`** | 자동으로 줄바꿈됨 (한 줄 독점) | 부모 너비의 100%를 기본 차지 | 상하좌우 마진/패딩 모두 정상 작동 | `div`, `p`, `h1~h6`, `ul`, `li` |
| **`inline`** | 줄바꿈 없이 가로 배치 | 콘텐츠 글자 크기만큼만 가로폭 차지 | **상하 마진/패딩이 레이아웃에 적용되지 않음** (가로 방향 패딩/마진만 가능) | `span`, `a`, `strong`, `em` |
| **`inline-block`** | 줄바꿈 없이 가로 배치 (inline의 흐름) | 원하는 대로 가로/세로 폭 지정 가능 | 상하좌우 마진/패딩 모두 정상 작동 | `input`, `button`, `select` (기본값) |
| **`none`** | 화면에서 완전히 삭제 | - | 공간 자체를 차지하지 않고 흔적 없이 사라짐 | (스크립트 제어용 혹은 모바일 메뉴 감춤 등) |

## 5. CSS 단위 체계 정리

* **`px` (Pixels)**: 절대적인 화소 단위로 크기가 항상 일정합니다. 해상도 왜곡이 없는 정밀 정렬에 유용하지만 사용자가 브라우저 글꼴 크기를 키워도 글씨가 커지지 않는 접근성 한계가 있습니다.
* **`%` (Percent)**: 부모 요소의 크기를 기준으로 계산되는 비례 단위입니다.
* **`em`**: **부모 요소의 폰트 크기**를 $1\text{em}$으로 계산하는 상대 단위입니다. (중첩 구조에서 폰트 크기 계산이 복잡해질 위험 존재)
* **`rem` (Root em)**: 최상위 `<html>` 요소의 폰트 크기(기본 브라우저 $16\text{px}$)를 $1\text{rem}$으로 고정 계산합니다. 
  * 예: $1.5\text{rem} = 16\text{px} \times 1.5 = 24\text{px}$
  * *권장*: 반응형 레이아웃 및 텍스트는 브라우저 글꼴 확대 옵션에 호응하기 위해 **`rem`**을 기본 단위로 사용합니다.
* **`vw` / `vh` (Viewport Width / Viewport Height)**: 사용자가 보는 웹 브라우저의 현재 창 크기를 100등분한 단위입니다. (예: `100vh`는 화면 세로 높이의 $100\%$ 꽉 차는 영역을 뜻함)

---

## 6. 타이포그래피 & 데코레이션 스타일

### 1) 텍스트 스타일 핵심 속성
* `font-family`: 글꼴 목록 설정. 마지막에는 항상 시스템 대체 글꼴 계열(`sans-serif` 등)을 선언해야 안심할 수 있습니다.
* `font-size`: 글씨 크기 (rem 권장).
* `font-weight`: 글씨 두께 (100 ~ 900 단위 또는 `normal`, `bold`).
* `line-height`: 줄간격. 단순히 `px`보다 글씨 크기 대비 배수(예: `1.5` ~ `1.6`)로 적는 것이 텍스트 겹침을 방지하는 표준 기법입니다.
* `text-align`: 텍스트 정렬 (`left`, `center`, `right`, `justify` 양끝정렬).
* `letter-spacing`: 글자 간의 간격(자간). 가독성 튜닝에 필수적입니다.

### 2) 배경 및 꾸밈
* `background-size`: 배경 이미지를 맞추는 기준입니다.
  * `cover`: 이미지 비율을 유지하며 요소 공간을 빈틈없이 꽉 채웁니다.
  * `contain`: 이미지가 잘리지 않는 선에서 요소 공간 내에 최대한 크게 맞춥니다.
* `border-radius`: 테두리 모서리를 둥글게 깎아줍니다. 둥근 사각형이나 원형(`50%`)을 만들 때 활용됩니다.
* `box-shadow`: 요소 섀도우(그림자) 효과를 줍니다.
  * `box-shadow: horizontal vertical blur spread color;` (예: `box-shadow: 2px 4px 10px rgba(0,0,0,0.15);`)

## 7. 핵심 키워드 정리
* **box-sizing**: 테두리 두께와 내부 안쪽 여백을 총 가로너비 계산에 편입시킬지 규정하는 속성.
* **rem/em 차이**: 기준이 문서 최상위 폰트크기(`html`)인지, 해당 요소 직속 부모 폰트크기인지에 따른 차이.
* **line-height**: 텍스트의 가독성 있는 행간 확보를 위한 높이 비율 속성.

# Flexbox 레이아웃 기초

## 1. 학습 목표
* 과거 테이블 마크업 및 float 기반 레이아웃 기법의 한계점과 부작용을 이해하고, 이를 극복하기 위해 등장한 1차원 레이아웃 모델 Flexbox의 출현 배경을 파악합니다.
* Flex 컨테이너(부모)와 Flex 아이템(자식) 간의 위계적 구조 관계를 이해합니다.
* 주축(Main Axis)과 교차축(Cross Axis)의 개념을 이해하고, `flex-direction`에 따라 두 축의 정렬 흐름이 동적으로 전환되는 원리를 파악합니다.
* 컨테이너 정렬 핵심 속성(`justify-content`, `align-items`, `flex-wrap`, `gap`)과 아이템 제어 속성(`flex-grow`, `flex-shrink`, `flex-basis`, `align-self`)의 의미와 역할을 숙지합니다.

---

## 2. Flexbox 등장 배경 (float 방식과의 비교)
기존에 웹 페이지의 요소를 가로 정렬하기 위해 주로 활용되던 방법은 `float: left;` 또는 `float: right;` 였습니다. 

* **기존 float 방식의 문제점**: 
  1. float 속성을 가진 자식 요소를 부모 요소가 제대로 인식하지 못해 **부모 영역 높이가 0으로 주저앉는 버그**가 지속적으로 발생했습니다. 이를 해결하기 위해 부모에 `clearfix`라는 가짜 해제용 가상 요소를 덧붙이거나 `overflow: hidden;`을 임시방편으로 걸어야만 했습니다.
  2. 세로 정렬(Vertical Center) 구현이 매우 까다로워 수동 픽셀값이나 `transform: translateY` 계산을 동원해야 했습니다.
  3. 형제 요소들 간의 높이를 동일하게 맞추는 등의 1차원 비율 분배 배치가 불가능에 가까웠습니다.

* **Flexbox의 혁신**:
  `display: flex;` 한 줄만 선언하면, 부모-자식 높이가 즉시 연동되고, 상하좌우 모든 정렬 및 간격 조절이 선언형 코드로 깔끔하게 완성됩니다.

---

## 3. 주축(Main Axis)과 교차축(Cross Axis) 개념
Flexbox 레이아웃 모델은 **주축(Main Axis)**과 **교차축(Cross Axis)**이라는 직교하는 두 축을 기준으로 삼습니다.

```
[flex-direction: row 인 경우 (기본값)]

           주축 (Main Axis): 좌 -> 우 (justify-content가 정렬 제어)
        ┌─────────────────────────────────────────────────────────┐
        │  [Item 1]    [Item 2]    [Item 3]                       │
  교차축│                                                         │
  (Cross│                                                         │
  Axis) │                                                         │
   ↓    │                                                         │
        └─────────────────────────────────────────────────────────┘
        (align-items 가 교차축 정렬 제어)
```

```
[flex-direction: column 인 경우]

        주축 (Main Axis): 위 -> 아래 (justify-content가 세로 정렬 제어)
        ┌──────────────────┐
        │   [Item 1]       │
        │                  │
        │   [Item 2]       │
        │                  │
  교차축│   [Item 3]       │ (Cross Axis: 좌 -> 우, align-items가 가로 정렬 제어)
   →    │                  │
        └──────────────────┘
```

> [!IMPORTANT]
> `flex-direction: column`으로 전환되면 주축이 세로로 바뀌고 교차축이 가로로 회전합니다. 이에 따라 `justify-content`는 세로(주축) 방향 정렬을, `align-items`는 가로(교차축) 방향 정렬을 조작하게 되므로 헷갈리지 않게 반드시 숙지해야 합니다.

---

## 4. Flex 컨테이너(부모)에 쓰는 핵심 속성

### 1) `flex-direction` (정렬 방향 설정)
* `row` (기본값): 왼쪽에서 오른쪽으로 가로 배치
* `row-reverse`: 오른쪽에서 왼쪽으로 가로 배치
* `column`: 위에서 아래로 세로 배치
* `column-reverse`: 아래에서 위로 세로 배치

### 2) `justify-content` (주축 정렬 방식)
* `flex-start` (기본값): 시작점 기준으로 요소를 몰아서 정렬
* `flex-end`: 끝점 기준으로 요소를 몰아서 정렬
* `center`: 주축 한가운데로 요소를 모아 정렬
* `space-between`: 첫 요소와 끝 요소를 양끝 벽에 고정하고 남은 요소들 간격을 동일하게 자동 배치
* `space-around`: 각 요소의 좌우에 동일한 마진 여백 부여 (요소 간 여백이 양끝 벽 여백의 2배가 됨)
* `space-evenly`: 양끝 벽과 요소들 사이의 모든 빈 여백 간격을 균등하게 배분

### 3) `align-items` (교차축 정렬 방식 - 한 줄 기준)
* `stretch` (기본값): 아이템에 높이(또는 너비) 지정이 없을 때 교차축 크기만큼 100% 늘려서 채움
* `flex-start`: 교차축 시작점(위쪽)에 배치
* `flex-end`: 교차축 끝점(아래쪽)에 배치
* `center`: 교차축 세로 한가운데 정렬 (수직 중앙 정렬의 해법)
* `baseline`: 폰트의 밑줄 기준선(Base Line)에 맞춰 텍스트 크기가 달라도 정렬

### 4) `flex-wrap` (줄바꿈 여부)
* `nowrap` (기본값): 부모 너비가 모자라도 찌그러져서 한 줄에 욱여넣어짐 (줄바꿈 안 됨)
* `wrap`: 부모 너비가 부족하면 자식 요소가 아래 줄로 자동으로 밀려 내려감
* `wrap-reverse`: 줄바꿈 시 위쪽 줄로 밀려 올라감

### 5) `gap` (아이템 간의 간격)
* `gap: 15px;` 형태로 선언하며, 아이템과 아이템 사이에만 고유한 여백을 만듭니다. (마진처럼 첫 요소나 끝 요소의 벽 쪽에는 불필요한 마진이 튀어나오지 않아 레이아웃 계산에 극히 편리함)

---

## 5. Flex 아이템(자식)에 쓰는 개별 제어 속성

### 1) `flex-grow` (팽창 비율)
* **정의**: 부모 컨테이너 내부에 남는 잉여 여백이 있을 때, 해당 자식이 여백을 나누어 먹으며 늘어날 비율을 의미합니다.
* **기본값**: `0` (여백이 남아도 스스로 늘어나지 않음)
* **비율**: A 아이템이 `1`, B 아이템이 `2`이면 남는 잉여 여백을 1:2의 지분 비율로 가져가 팽창합니다.

### 2) `flex-shrink` (수축 비율)
* **정의**: 컨테이너의 크기가 자식들의 총합보다 작아져 공간이 부족할 때, 자식을 찌그러뜨려 축소할 비율을 설정합니다.
* **기본값**: `1` (공간 부족 시 균등 비율로 축소됨)
* **설정**: `0`으로 지정하면 부모 창이 좁아져도 자신의 절대 크기 크기를 훼손당하지 않고 원래 크기를 유지합니다.

### 3) `flex-basis` (기본 시작 크기)
* **정의**: `flex-grow`나 `flex-shrink`에 의해 팽창/수축하기 전, 요소가 가지는 고유의 기본 시작 픽셀 크기 값을 가리킵니다. (`width`와 유사한 역할이지만 Flexbox 환경에서는 `flex-basis`가 우선 적용됨)
* **기본값**: `auto`

### 4) `flex` 축약형 속성
* `flex: grow shrink basis;` 의 형태로 한 번에 축약 선언합니다.
* **예시**: `flex: 1;` -> `flex-grow: 1; flex-shrink: 1; flex-basis: 0%;` 를 생략 선언한 것으로, 남는 모든 여백 공간을 다른 형제와 균등하게 나눠 가지며 부피를 꽉 채우겠다는 핵심 실무 관용구입니다.

### 5) `align-self` (개별 정렬)
* 부모가 정해준 전체 `align-items` 규칙을 무시하고, **해당 개별 자식 아이템만** 홀로 교차축에서 다른 방향으로 탈선하여 정렬되도록 지정합니다.

---

## 6. 핵심 키워드 정리
* **주축/교차축**: `flex-direction`에 종속되어 정렬의 척도를 가르는 교차선.
* **justify-content vs align-items**: 각각 주축 방향 정렬과 교차축 방향 정렬을 처리하는 속성.
* **flex-wrap**: 자식 요소가 부모 범위를 탈출하려 할 때 아래 라인으로 자동 갱신할지 여부.

# Flexbox 실전 레이아웃 구현

## 1. 학습 목표
* 헤더, 가변형 카드 목록, 멀티 컬럼 푸터 등 현대 웹 페이지의 빈출 레이아웃 구성요소를 Flexbox로 완전히 조작할 수 있습니다.
* 부모도 Flexbox이고, 그 내부의 자식 요소 역시 Flexbox 컨테이너가 되는 **Flexbox 중첩(Nesting)** 설계의 원리를 파악합니다.
* `flex: 1` 선언이 가지는 속성값의 기하학적 의미(`flex-grow:1`, `flex-shrink:1`, `flex-basis:0%`)를 이해하고, 이를 이용해 남은 높이/너비를 꽉 채우거나 버튼을 박스 하단에 밀어붙이는 실무 패턴을 적용합니다.
* `flex-wrap: wrap`과 퍼센티지 너비 또는 `flex-basis`를 이용해 화면 크기 감소 시 자연스럽게 카드가 다음 줄로 넘어가는 반응형 기초 배열을 흉내 냅니다.

---

## 2. 실무 레이아웃의 3대 핵심 패턴

### 1) 헤더 (GNB 양끝 배치)
* **특징**: 왼쪽의 로고와 오른쪽의 내비게이션 메뉴 묶음을 양끝단 벽에 완전히 밀어붙이고 세로를 한가운데 맞추는 패턴입니다.
* **해법**:
  ```css
  .header-container {
    display: flex;
    justify-content: space-between;
    align-items: center;
  }
  ```

### 2) 가로 세로 비율 고정 카드 갤러리 (Card Grid)
* **특징**: 여러 장의 카드를 가로 방향으로 나열하다가 화면 가로 폭이 부족하면 카드를 깨뜨리지 않고 다음 라인으로 줄바꿈합니다.
* **해법**:
  ```css
  .card-container {
    display: flex;
    flex-wrap: wrap; /* 줄바꿈 허용 */
    gap: 20px; /* 카드간 고정 여백 */
  }
  .card-item {
    flex: 1 1 calc(33.33% - 20px); /* 3열 배치 기본 시작, 최소 부피 보장 */
    min-width: 280px; /* 모바일에서 찌그러지지 않고 1열로 떨어지게 하는 방어막 */
  }
  ```

### 3) 3컬럼 푸터 (Multi-column Footer)
* **특징**: 하단 꼬리말 영역에서 회사 소개, 고객 센터 링크, 소셜 미디어 링크 등 3개 이상의 기둥(Column)을 가로로 정렬하며, 모바일 화면에서는 세로로 차곡차곡 쌓이게 만듭니다.
* **해법**: `flex-direction: row`에서 모바일 미디어 진입 시 `flex-direction: column`으로 전환하는 기본 발상을 활용합니다.

---

## 3. Flexbox 중첩 (Nesting) 패턴
Flexbox 레이아웃을 완성도 높게 짜기 위해서는 중첩 기법을 이해해야 합니다.

```
┌─────────────────────────────────────────────────────────────────┐ <- 1차 부모 (display: flex)
│  ┌──────────────────────┐  ┌──────────────────────┐             │
│  │ [Card 1 (Flex Item)] │  │ [Card 2 (Flex Item)] │             │
│  │                      │  │                      │             │
│  │  ┌────────────────┐  │  │                      │             │
│  │  │  [Card Image]  │  │  │                      │             │
│  │  └────────────────┘  │  │                      │             │
│  │                      │  │                      │             │
│  │  ┌────────────────┐  │  │                      │             │
│  │  │  [Card Title]  │  │  │                      │             │
│  │  └────────────────┘  │  │                      │             │
│  │                      │  │                      │             │
│  │  ┌────────────────┐  │  │                      │             │
│  │  │  [Card Button] │  │  │                      │             │ <- 2차 자식이자
│  │  └────────────────┘  │  │                      │             │    세부 부모 (display: flex,
│  └──────────────────────┘  └──────────────────────┘             │    flex-direction: column)
└─────────────────────────────────────────────────────────────────┘
```

* **바깥쪽 부모**: 카드 리스트 전체를 정렬하기 위해 `display: flex;` 및 `flex-wrap: wrap;`을 부여합니다.
* **안쪽 자식 (카드 개별 박스)**: 그 자체는 바깥 부모의 아이템이지만, 카드 내부의 이미지, 텍스트, 하단 고정 버튼을 세로 일직선 상에 균일한 비율로 배치하기 위해 내부에도 `display: flex;` 및 `flex-direction: column;`을 부여합니다.

---

## 4. 자주 하는 실수 및 실무 팁

### 1) `flex: 1`의 정확한 정체
* `flex: 1;`은 `flex-grow: 1; flex-shrink: 1; flex-basis: 0%;`와 동일합니다.
* 여기서 `flex-basis: 0%;`로 셋팅되기 때문에 자식 요소의 기존 콘텐츠 고유 크기(`width`)를 완전히 무시하고, **오로지 부모가 가진 전체 남는 가로폭 여백만을 수학적으로 1:1 정확히 나누어 가지게 만듭니다.**
* 만약 `flex-grow: 1;`만 지정하고 `flex-basis: auto;`로 두면 각 자식이 담고 있는 글자 길이에 비례해 카드 가로 크기가 짝짝이가 되는 버그가 일어납니다. 따라서 균등 너비 레이아웃에는 반드시 축약형인 `flex: 1;`을 사용합니다.

### 2) 버튼을 카드의 맨 아래로 완벽히 붙이기 (`margin-top: auto`)
* 카드 내부의 소개글 텍스트 분량이 카드마다 다른 경우, 카드 전체의 높이는 똑같아지더라도 버튼의 위치가 글자 줄 수에 맞춰 삐뚤빼뚤 춤추게 됩니다.
* **해결법**: 카드 개별 박스를 `flex-direction: column`으로 선언하고, 가장 밑에 있는 버튼 요소에 **`margin-top: auto;`**를 선언해줍니다.
* **원리**: Flexbox 환경에서 특정 자식에 `auto` 마진을 주면, 주축 방향의 남은 모든 빈 여백을 해당 마진이 혼자서 전부 흡수해버립니다. 따라서 버튼 위의 빈 공간을 최대로 벌려서 버튼이 카드의 가장 아래쪽 바닥선에 자석처럼 찰칵 고정되도록 만들어 줍니다.

---

## 5. 핵심 키워드 정리
* **flex: 1**: 콘텐츠 너비를 무시하고 남은 공간을 균등 지분율로 채워 가로폭을 분할하는 축약 명령.
* **margin-top: auto (Flexbox 내)**: 세로형 Flex 내부에서 특정 아이템을 바닥 끝단으로 강제 이동시키는 실무 테크닉.
* **Flex 중첩**: 레이아웃 전체 틀과 개별 상세 정보 영역을 각각 1차원 Flex로 상호 제어하는 방식.

# CSS Grid 레이아웃

## 학습 목표



- CSS Grid의 핵심 개념(컨테이너, 아이템, 트랙, 셀, 갭)을 설명할 수 있다.
- `grid-template-columns`, `grid-template-rows`로 격자를 자유롭게 정의할 수 있다.
- `fr`, `repeat()`, `minmax()`, `auto-fit/auto-fill`을 활용한 유연한 레이아웃을 구성할 수 있다.
- `grid-template-areas`로 직관적인 대시보드 레이아웃을 설계할 수 있다.
- `grid-column`, `grid-row`로 아이템을 특정 셀에 배치하거나 병합할 수 있다.
- Flexbox와 Grid의 차이를 이해하고 상황에 맞게 선택할 수 있다.

## 1. Flexbox(1차원) vs Grid(2차원)

CSS 레이아웃의 양대 산맥인 Flexbox와 Grid는 **적용 차원**이 다릅니다.  
"무엇을 쓸까?" 헷갈릴 때 아래 표를 기준으로 판단하세요.

| 비교 항목 | Flexbox | Grid |
|---|---|---|
| **차원** | 1차원 (행 또는 열 중 하나) | 2차원 (행과 열 동시) |
| **배치 방향** | 단방향 (가로 또는 세로) | 양방향 (가로 + 세로) |
| **주요 사용처** | 네비게이션 메뉴, 카드 정렬, 버튼 그룹 | 페이지 레이아웃, 대시보드, 갤러리 |
| **아이템 크기 제어** | 콘텐츠 기반 (flex-grow/shrink) | 명시적 트랙 기반 (fr, px, %) |
| **정렬 기준** | 주축(main axis) / 교차축(cross axis) | 행 축 / 열 축 |
| **겹침 배치** | 불가 (기본) | 가능 (grid-column/row 으로) |
| **브라우저 지원** | IE 10+ (구형도 가능) | IE 11+ (일부 제한) |

###  선택 기준 공식

```
컴포넌트 단위의 1방향 배열    → Flexbox
페이지 전체 혹은 복잡한 2D 배치 → Grid
두 기술을 함께 쓰는 것이 가장 강력합니다 (Grid 외곽 + Flex 내부)
```

---

## 2. Grid 기본 개념

### 2.1 용어 정리

```
┌──────────────────────────────────────┐
│          Grid Container              │
│  ┌────────┬────────┬────────┐        │
│  │  Cell  │  Cell  │  Cell  │ ← Row  │
│  ├────────┼────────┼────────┤  Track │
│  │  Cell  │  Cell  │  Cell  │        │
│  └────────┴────────┴────────┘        │
│    ↑ Column Track                    │
│         ↕ Gap (row-gap)              │
│       ↔ Gap (column-gap)             │
└──────────────────────────────────────┘
```

| 용어 | 설명 |
|---|---|
| **Grid Container** | `display: grid`가 선언된 부모 요소 |
| **Grid Item** | 컨테이너의 직계 자식 요소 |
| **Grid Line** | 격자를 나누는 선 (번호로 참조: 1, 2, 3...) |
| **Grid Cell** | 행선과 열선으로 둘러싸인 하나의 격자 칸 |
| **Grid Track** | 두 선 사이의 공간 (행 트랙 또는 열 트랙) |
| **Grid Area** | 하나 이상의 셀로 이루어진 직사각형 영역 |
| **Gap** | 트랙과 트랙 사이의 여백 |

### 2.2 그리드 라인 번호 시스템

```
    1    2    3    4   ← Column Lines
  ┌────┬────┬────┐
1 │    │    │    │
  ├────┼────┼────┤
2 │    │    │    │
  ├────┼────┼────┤
3 │    │    │    │
  └────┴────┴────┘
↑ Row Lines
```

- 열 라인: 왼쪽부터 1, 2, 3, 4 (n열이면 n+1개)
- 행 라인: 위쪽부터 1, 2, 3, 4 (n행이면 n+1개)
- 음수 인덱스: -1(마지막), -2(끝에서 두 번째)...

## 3. Grid 컨테이너 만들기

```css
.container {
  display: grid;        /* 블록 레벨 그리드 */
  display: inline-grid; /* 인라인 레벨 그리드 */
}
```

`display: grid`를 선언하는 순간 직계 자식들이 **Grid Item**이 됩니다.

---

## 4. grid-template-columns / grid-template-rows

격자의 열과 행 크기를 정의하는 핵심 속성입니다.

### 4.1 단위 종류

| 단위 | 설명 | 예시 |
|---|---|---|
| `px` | 고정 픽셀 크기 | `200px` |
| `%` | 컨테이너 너비 기준 비율 | `50%` |
| `fr` | 남은 공간의 비율 (**fraction**) | `1fr 2fr` |
| `auto` | 콘텐츠 크기에 맞게 자동 조절 | `auto` |
| `min-content` | 콘텐츠의 최소 크기 | `min-content` |
| `max-content` | 콘텐츠의 최대 크기 | `max-content` |

#### `fr` 단위 동작 원리

```css
.container {
  grid-template-columns: 1fr 2fr 1fr;
  /* 전체 너비를 4등분 (1+2+1=4)
     첫 번째 열: 1/4
     두 번째 열: 2/4 (가장 넓음)
     세 번째 열: 1/4         */
}
```

```
┌────────┬────────────────┬────────┐
│  1fr   │      2fr       │  1fr   │
│ (25%)  │     (50%)      │ (25%)  │
└────────┴────────────────┴────────┘
```

> **💡 `fr` vs `%` 차이**  
> `%`는 gap을 무시하고 계산, `fr`은 gap을 제외한 남은 공간을 기준으로 계산합니다.  
> 따라서 gap이 있는 경우 `fr`이 더 정확합니다.

### 4.2 repeat() 함수

같은 패턴을 반복할 때 사용합니다.

```css
/* 기본 문법 */
grid-template-columns: repeat(횟수, 크기);

/* 예시들 */
grid-template-columns: repeat(3, 1fr);
/* 동일: 1fr 1fr 1fr */

grid-template-columns: repeat(4, 200px);
/* 동일: 200px 200px 200px 200px */

grid-template-columns: repeat(3, 1fr 2fr);
/* 동일: 1fr 2fr 1fr 2fr 1fr 2fr → 6열 */

/* 혼합 사용 */
grid-template-columns: 250px repeat(3, 1fr);
/* 첫 열 250px 고정, 나머지 3열 균등 분배 */
```

### 4.3 minmax() 함수

트랙의 **최소**와 **최대** 크기를 동시에 지정합니다.

```css
/* 문법 */
minmax(최솟값, 최댓값)

/* 예시 */
grid-template-columns: minmax(200px, 1fr);
/* 최소 200px, 최대는 남은 공간 전체 */

grid-template-columns: repeat(3, minmax(150px, 1fr));
/* 3열, 각 열은 최소 150px ~ 최대 균등 */

/* 행에도 적용 가능 */
grid-template-rows: minmax(100px, auto);
/* 최소 100px, 콘텐츠가 많으면 자동 확장 */
```

### 4.4 auto-fit vs auto-fill

`repeat()`와 함께 사용하여 **컨테이너 너비에 따라 열 수를 자동으로 결정**합니다.

```css
/* auto-fill: 빈 트랙도 공간 차지 */
grid-template-columns: repeat(auto-fill, minmax(200px, 1fr));

/* auto-fit: 빈 트랙을 접어서 아이템이 늘어남 */
grid-template-columns: repeat(auto-fit, minmax(200px, 1fr));
```

#### 시각적 차이 (아이템 3개, 컨테이너 넓을 때)

```
auto-fill (빈 칸 유지):
┌─────┬─────┬─────┬─────┬─────┐
│  1  │  2  │  3  │     │     │  ← 빈 트랙 존재
└─────┴─────┴─────┴─────┴─────┘

auto-fit (빈 칸 제거):
┌───────────┬───────────┬───────────┐
│     1     │     2     │     3     │  ← 아이템이 공간 채움
└───────────┴───────────┴───────────┘
```

| 상황 | 권장 값 |
|---|---|
| 아이템 고정 크기 유지 | `auto-fill` |
| 아이템이 공간을 채워야 할 때 | `auto-fit` |
| 카드 그리드 (일반적) | `auto-fit` + `minmax()` |

## 5. gap (여백)

```css
.container {
  gap: 20px;             /* row-gap과 column-gap 동시 설정 */
  gap: 20px 40px;        /* row-gap: 20px, column-gap: 40px */
  
  row-gap: 20px;         /* 행 사이 여백 */
  column-gap: 40px;      /* 열 사이 여백 */
}
```

> **⚠️ 구형 속성명**  
> 과거에는 `grid-gap`, `grid-row-gap`, `grid-column-gap`을 사용했으나  
> 현재 표준은 `gap`, `row-gap`, `column-gap`입니다.

## 6. grid-template-areas — 영역 이름으로 레이아웃 설계

가장 직관적인 레이아웃 방법입니다. CSS 안에서 레이아웃을 **ASCII 아트**처럼 그릴 수 있습니다.

### 6.1 기본 문법

```css
.container {
  display: grid;
  grid-template-areas:
    "영역명1 영역명2 영역명3"  /* 1행 */
    "영역명4 영역명5 영역명6"  /* 2행 */
    "영역명7 영역명8 영역명9"; /* 3행 */
}
```

### 6.2 대시보드 레이아웃 예제

#### 레이아웃 설계도

```
┌─────────────────────────────────┐
│            HEADER               │
├──────────┬──────────┬───────────┤
│          │          │           │
│   NAV    │   MAIN   │   ASIDE   │
│ (사이드바) │ (콘텐츠) │  (위젯)   │
│          │          │           │
├──────────┴──────────┴───────────┤
│            FOOTER               │
└─────────────────────────────────┘
```

#### CSS 코드

```css
.dashboard {
  display: grid;
  grid-template-columns: 200px 1fr 250px;
  grid-template-rows: 60px 1fr 50px;
  grid-template-areas:
    "header header header"
    "nav    main   aside"
    "footer footer footer";
  min-height: 100vh;
  gap: 10px;
}

/* 각 아이템에 grid-area 할당 */
.header { grid-area: header; }
.nav    { grid-area: nav;    }
.main   { grid-area: main;   }
.aside  { grid-area: aside;  }
.footer { grid-area: footer; }
```

#### HTML 구조

```html
<div class="dashboard">
  <header class="header">헤더</header>
  <nav class="nav">사이드 내비게이션</nav>
  <main class="main">메인 콘텐츠</main>
  <aside class="aside">위젯 영역</aside>
  <footer class="footer">푸터</footer>
</div>
```

### 6.3 영역 병합하기

같은 이름을 반복하면 자동으로 병합됩니다.

```css
grid-template-areas:
  "header header header"  /* header가 3칸 차지 */
  "nav    main   main"    /* main이 2칸 차지 */
  "nav    footer footer"; /* nav가 2행 차지 */
```

```
┌──────┬──────┬──────┐
│         header      │
├──────┬──────┬──────┤
│      │             │
│ nav  │    main     │
│      ├──────┬──────┤
│      │    footer   │
└──────┴──────┴──────┘
```

> **규칙**: 같은 이름의 셀은 반드시 **직사각형** 모양을 이루어야 합니다.  
> 빈 칸으로 두려면 `.` (마침표)을 사용합니다.

## 7. grid-area로 아이템 배치

### 7.1 이름으로 배치 (grid-template-areas와 함께)

```css
.item { grid-area: main; } /* 'main' 영역에 배치 */
```

### 7.2 라인 번호로 배치 (축약형)

```css
/* grid-area: 행시작 / 열시작 / 행끝 / 열끝 */
.item {
  grid-area: 1 / 2 / 3 / 4;
  /* row-start: 1, column-start: 2
     row-end: 3,   column-end: 4  */
}
```

## 8. grid-column / grid-row — 개별 아이템 배치

### 8.1 기본 문법

```css
.item {
  /* grid-column: 시작라인 / 끝라인 */
  grid-column: 1 / 3;   /* 1번 열선 ~ 3번 열선 (2열 차지) */
  grid-row: 1 / 2;      /* 1번 행선 ~ 2번 행선 (1행 차지) */
}
```

### 8.2 span 키워드

```css
.item {
  grid-column: 1 / span 2;  /* 1번 열선부터 2칸 차지 */
  grid-column: span 3;       /* 현재 위치에서 3칸 차지 */
  
  grid-row: 2 / span 3;     /* 2번 행선부터 3행 차지 */
}
```

### 8.3 음수 라인 번호

```css
.item {
  grid-column: 1 / -1;   /* 첫 열부터 마지막 열까지 (전체 너비) */
  grid-row: -3 / -1;     /* 끝에서 세 번째 행부터 마지막 행까지 */
}
```

### 8.4 배치 예시

```
3×3 그리드에서:
┌────┬────┬────┐
│ A  │    B   │  A: grid-column: 1/2
├────┼────┼────┤  B: grid-column: 2/4
│         │ D │  C: grid-column: 1/3, grid-row: 2/4
│    C    ├────┤  D: grid-column: 3/4, grid-row: 2/3
│         │ E │  E: grid-column: 3/4, grid-row: 3/4
└─────────┴────┘
```

```css
.A { grid-column: 1 / 2; grid-row: 1 / 2; }
.B { grid-column: 2 / 4; grid-row: 1 / 2; }
.C { grid-column: 1 / 3; grid-row: 2 / 4; }
.D { grid-column: 3 / 4; grid-row: 2 / 3; }
.E { grid-column: 3 / 4; grid-row: 3 / 4; }
```

## 9. 정렬 속성

### 9.1 컨테이너 기준 (모든 아이템에 적용)

| 속성 | 기본값 | 설명 |
|---|---|---|
| `justify-items` | `stretch` | 아이템을 **열 축(수평)** 기준으로 정렬 |
| `align-items` | `stretch` | 아이템을 **행 축(수직)** 기준으로 정렬 |
| `place-items` | `stretch` | `align-items` / `justify-items` 축약형 |

```css
.container {
  justify-items: start | end | center | stretch;
  align-items:   start | end | center | stretch;
  
  /* place-items: align justify; */
  place-items: center;         /* 가로세로 모두 가운데 */
  place-items: start end;      /* 세로: start, 가로: end */
}
```

### 9.2 개별 아이템 기준 (특정 아이템에만 적용)

| 속성 | 설명 |
|---|---|
| `justify-self` | 해당 아이템만 수평 정렬 |
| `align-self` | 해당 아이템만 수직 정렬 |
| `place-self` | 축약형 |

```css
.special-item {
  justify-self: end;     /* 이 아이템만 오른쪽 정렬 */
  align-self: center;    /* 이 아이템만 수직 가운데 */
  place-self: center end;
}
```

### 9.3 justify-content / align-content (그리드 전체 정렬)

그리드 트랙 전체가 컨테이너보다 작을 때 트랙 묶음 전체를 정렬합니다.

```css
.container {
  justify-content: start | end | center | space-between | space-around | space-evenly;
  align-content:   start | end | center | space-between | space-around | space-evenly;
}
```

## 10. 자동 배치 (Implicit Grid)

명시적으로 지정하지 않은 아이템은 **암묵적 그리드**에 자동 배치됩니다.

```css
.container {
  grid-auto-flow: row;    /* 기본: 행 방향으로 자동 배치 */
  grid-auto-flow: column; /* 열 방향으로 자동 배치 */
  grid-auto-flow: dense;  /* 빈 공간을 채우는 밀집 배치 */
  
  grid-auto-rows: 150px;     /* 암묵적 행의 기본 크기 */
  grid-auto-columns: 100px;  /* 암묵적 열의 기본 크기 */
}
```

## 1. 반응형 웹 디자인의 필요성

### 1.1 모바일 트래픽 현황

| 연도 | 모바일 웹 트래픽 비율 | 데스크탑 비율 |
|------|----------------------|---------------|
| 2015 | 35.1% | 64.9% |
| 2018 | 52.4% | 47.6% |
| 2021 | 57.0% | 43.0% |
| 2024 | **60% 이상** | 40% 미만 |

> ** 결론**: 전 세계 웹 트래픽의 60% 이상이 모바일에서 발생합니다.  
> 고정 너비 레이아웃은 모바일에서 가로 스크롤이 생기거나 글자가 너무 작아집니다.

### 1.2 반응형 디자인 핵심 원칙 3가지

```
1. 유동 그리드(Fluid Grid)    → px 대신 %, fr, em, rem 사용
2. 유연한 이미지(Flexible Image) → max-width: 100%로 이미지 제한
3. 미디어 쿼리(Media Query)   → 화면 크기에 따른 CSS 규칙 분기
```

## 2. meta viewport 완전 이해

### 2.1 viewport란?

**뷰포트(Viewport)**는 웹 페이지가 표시되는 **브라우저의 보이는 영역**입니다.

```
데스크탑: viewport 너비 ≈ 1200~1920px
태블릿:   viewport 너비 ≈ 768~1024px
모바일:   viewport 너비 ≈ 320~480px
```

### 2.2 viewport 메타 태그

```html
<meta name="viewport" content="width=device-width, initial-scale=1.0">
```

| 속성값 | 의미 |
|--------|------|
| `width=device-width` | 뷰포트 너비를 디바이스의 실제 화면 너비로 설정 |
| `initial-scale=1.0` | 페이지 로드 시 초기 확대/축소 배율을 100%로 설정 |

> **⚠️ 이 태그가 없으면?**  
> 모바일 브라우저는 페이지를 데스크탑용(약 980px)으로 렌더링한 후 축소합니다.  
> 결과적으로 글씨가 매우 작아지고 미디어 쿼리도 제대로 작동하지 않습니다.

### 2.3 추가 옵션들

```html
<!-- 사용자 확대/축소 비허용 (접근성 이슈로 비권장) -->
<meta name="viewport" content="width=device-width, initial-scale=1.0, user-scalable=no">

<!-- 최대 확대 배율 제한 -->
<meta name="viewport" content="width=device-width, initial-scale=1.0, maximum-scale=2.0">
```

## 3. 미디어 쿼리 문법 완전 정리

### 3.1 기본 문법

```css
@media [미디어 타입] and ([조건]) {
  /* 조건이 참일 때 적용되는 CSS */
}
```

### 3.2 미디어 타입

| 타입 | 설명 | 주요 사용 |
|------|------|-----------|
| `all` | 모든 디바이스 (기본값) | 일반적인 스타일 |
| `screen` | 화면 (모니터, 폰, 태블릿) | 반응형 레이아웃 |
| `print` | 인쇄 미리보기 및 인쇄 | 인쇄용 CSS |
| `speech` | 스크린 리더 (음성) | 접근성 |

```css
/* 인쇄 시 불필요한 요소 숨기기 */
@media print {
  .nav, .ads, .footer { display: none; }
  body { font-size: 12pt; }
}
```

### 3.3 주요 미디어 특성(조건)

| 특성 | 설명 | 예시 |
|------|------|------|
| `min-width` | 뷰포트 너비가 이상일 때 | `(min-width: 768px)` |
| `max-width` | 뷰포트 너비가 이하일 때 | `(max-width: 767px)` |
| `min-height` | 뷰포트 높이가 이상일 때 | `(min-height: 600px)` |
| `orientation` | 화면 방향 | `(orientation: landscape)` |
| `hover` | 포인터 기기 호버 가능 여부 | `(hover: hover)` |
| `prefers-color-scheme` | 시스템 다크/라이트 모드 | `(prefers-color-scheme: dark)` |

### 3.4 논리 연산자

#### `and` — 두 조건 모두 만족

```css
/* 768px 이상이면서 1024px 이하인 경우 (태블릿 전용) */
@media screen and (min-width: 768px) and (max-width: 1024px) {
  .container { padding: 0 30px; }
}
```

#### `,` (쉼표) — OR 조건 (둘 중 하나 만족)

```css
/* 작은 화면 또는 인쇄 시 */
@media screen and (max-width: 480px), print {
  .sidebar { display: none; }
}
```

#### `not` — 조건 부정

```css
/* screen이 아닌 디바이스에서 */
@media not screen {
  body { font-family: serif; }
}
```

### 3.5 min-width vs max-width

이것이 모바일 퍼스트 vs 데스크탑 퍼스트의 핵심 차이입니다.

```css
/* ── min-width: 모바일 퍼스트 ──
   "이 너비 이상이면 적용" → 작은 화면이 기준 */
.container { width: 100%; }            /* 기본 (모바일) */
@media (min-width: 768px)  { .container { width: 750px; } }  /* 태블릿↑ */
@media (min-width: 1024px) { .container { width: 960px; } }  /* 데스크탑↑ */

/* ── max-width: 데스크탑 퍼스트 ──
   "이 너비 이하이면 적용" → 큰 화면이 기준 */
.container { width: 960px; }           /* 기본 (데스크탑) */
@media (max-width: 1023px) { .container { width: 750px; } }  /* 태블릿↓ */
@media (max-width: 767px)  { .container { width: 100%; } }   /* 모바일↓ */
```

## 4. 모바일 퍼스트 vs 데스크탑 퍼스트

### 4.1 전략별 장단점 비교

| 항목 | 모바일 퍼스트 | 데스크탑 퍼스트 |
|------|--------------|----------------|
| **기본 CSS** | 작은 화면 기준 | 큰 화면 기준 |
| **미디어쿼리 방향** | `min-width` (점차 확장) | `max-width` (점차 축소) |
| **성능** |  모바일 불필요한 CSS 없음 |  모바일에서 데스크탑 CSS 다운로드 |
| **SEO** |  Google 모바일 우선 색인 |  불리할 수 있음 |
| **개발 순서** | 핵심 기능 먼저 설계 | 풍부한 UI 먼저 설계 |
| **실무 권장** |  **권장** | △ 레거시 프로젝트에서 사용 |

### 4.2 코드 비교

```css
/* ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
   모바일 퍼스트 (권장)
   기본 = 모바일, min-width로 점점 확장
   ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━ */
.features {
  display: grid;
  grid-template-columns: 1fr;  /* 모바일: 1열 */
}

@media (min-width: 768px) {    /* 태블릿 이상 */
  .features {
    grid-template-columns: repeat(2, 1fr);  /* 2열 */
  }
}

@media (min-width: 1024px) {   /* 데스크탑 이상 */
  .features {
    grid-template-columns: repeat(3, 1fr);  /* 3열 */
  }
}


/* ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
   데스크탑 퍼스트 (비권장)
   기본 = 데스크탑, max-width로 점점 축소
   ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━ */
.features {
  display: grid;
  grid-template-columns: repeat(3, 1fr);  /* 데스크탑: 3열 */
}

@media (max-width: 1023px) {   /* 태블릿 이하 */
  .features {
    grid-template-columns: repeat(2, 1fr);  /* 2열 */
  }
}

@media (max-width: 767px) {    /* 모바일 이하 */
  .features {
    grid-template-columns: 1fr;  /* 1열 */
  }
}
```

> ** 권장** : 모바일 퍼스트를 사용하세요.  
> CSS는 위에서 아래로 적용되므로, 작은 화면 기본 스타일 → `min-width`로 확장하는 방식이  
> 코드 관리와 성능 모두에서 유리합니다.

## 5. 브레이크포인트(Breakpoint) 기준표

브레이크포인트는 레이아웃이 변경되는 뷰포트 너비 기준점입니다.

| 브레이크포인트 | 대상 기기 | 미디어 쿼리 | 일반적 레이아웃 |
|---------------|-----------|------------|----------------|
| `< 576px` | 소형 폰 (iPhone SE 등) | 기본 CSS | 1열, 전체 너비 |
| `≥ 576px` | 중형 폰 (갤럭시, 아이폰) | `@media (min-width: 576px)` | 1~2열 |
| `≥ 768px` | 태블릿 (iPad mini 등) | `@media (min-width: 768px)` | 2열, nav 표시 |
| `≥ 1024px` | 소형 노트북 | `@media (min-width: 1024px)` | 3열, 사이드바 |
| `≥ 1200px` | 데스크탑, 모니터 | `@media (min-width: 1200px)` | 최대 너비 컨테이너 |

```
320px  576px    768px      1024px        1200px
  |─────|─────────|──────────|─────────────|──────→
  폰     폰(중형)   태블릿      노트북         데스크탑
  1열    1열       2열         3열            3열+여백
```

## 6. 반응형 전환 패턴

### 6.1 Grid 컬럼 수 변경

```css
/* 카드 그리드: 1→2→3열 */
.card-grid {
  display: grid;
  grid-template-columns: 1fr;          /* 모바일: 1열 */
  gap: 16px;
}

@media (min-width: 768px) {
  .card-grid { grid-template-columns: repeat(2, 1fr); }  /* 태블릿: 2열 */
}

@media (min-width: 1024px) {
  .card-grid { grid-template-columns: repeat(3, 1fr); }  /* 데스크탑: 3열 */
}
```

### 6.2 flex-direction 전환

```css
/* 네비게이션: 세로 → 가로 */
.nav-list {
  display: flex;
  flex-direction: column;  /* 모바일: 세로 나열 */
  gap: 8px;
}

@media (min-width: 768px) {
  .nav-list {
    flex-direction: row;   /* 데스크탑: 가로 나열 */
    gap: 24px;
  }
}
```

### 6.3 이미지 반응형

```css
img {
  max-width: 100%;  /* 부모 요소를 초과하지 않도록 제한 */
  height: auto;     /* 비율을 유지하며 높이 자동 조절 */
  display: block;   /* 이미지 하단 공백 제거 */
}
```

### 6.4 폰트 크기 조정

```css
/* 클램핑: min ~ max 사이에서 유동적으로 크기 변경 */
h1 {
  font-size: clamp(1.8rem, 4vw, 3.5rem);
  /*         최소      권장    최대   */
}

/* 미디어 쿼리 방식 */
h1 { font-size: 1.8rem; }                        /* 모바일 */
@media (min-width: 768px)  { h1 { font-size: 2.4rem; } }  /* 태블릿 */
@media (min-width: 1024px) { h1 { font-size: 3.2rem; } }  /* 데스크탑 */
```

### 6.5 컨테이너 최대 너비 패턴

```css
.container {
  width: 100%;           /* 모바일: 전체 너비 */
  max-width: 1200px;     /* 데스크탑: 최대 1200px */
  margin: 0 auto;        /* 가운데 정렬 */
  padding: 0 16px;       /* 양옆 기본 여백 */
}

@media (min-width: 768px) {
  .container { padding: 0 24px; }
}

@media (min-width: 1024px) {
  .container { padding: 0 40px; }
}
```

## 7. Flex + Grid 혼합 설계 전략

실무에서는 Flexbox와 Grid를 **목적에 따라 혼합**하여 사용합니다.

```
페이지 레벨 레이아웃 (전체 구조)    → Grid
컴포넌트 내부 정렬 (버튼, 메뉴 등) → Flexbox
카드 갤러리 (2D 배치)               → Grid
단순 수평/수직 정렬                  → Flexbox
```

```css
/* Grid로 전체 레이아웃 설계 */
.page {
  display: grid;
  grid-template-areas:
    "header"
    "main"
    "footer";
}

/* Flex로 헤더 내부 정렬 */
.header {
  display: flex;
  justify-content: space-between;  /* 로고와 메뉴를 양 끝으로 */
  align-items: center;
}

/* Grid로 피처 섹션 카드 배치 */
.features {
  display: grid;
  grid-template-columns: repeat(auto-fit, minmax(280px, 1fr));
}

/* Flex로 카드 내부 요소 정렬 */
.card {
  display: flex;
  flex-direction: column;
  gap: 12px;
}
```

## 8. 종합 프로젝트 구조

이번 실습에서 구현할 랜딩 페이지의 전체 구조입니다.

```
index.html
├── <header>         ← 헤더 + 네비게이션 (Flexbox, sticky)
│   └── <nav>        ← 메뉴 (모바일: 세로 / 데스크탑: 가로)
├── <section#hero>   ← 히어로 배너 (Flex 중앙 정렬, 전체 화면 높이)
├── <section#features> ← 특징 카드 (Grid: 1→2→3열 반응형)
├── <section#about>  ← 회사 소개 (Grid: 1→2열 반응형)
├── <section#gallery> ← 갤러리 (Flexbox wrap)
├── <section#contact> ← 문의 폼 (CSS 폼 스타일링)
└── <footer>         ← 푸터 (Flexbox: 모바일 세로/데스크탑 가로)
```

### 차시별 기술 활용 현황

| 구역 | 적용 기술 | 학습 차시 |
|------|-----------|-----------|
| `<header>` | display:flex, sticky position | 6차시 |
| `<nav>` | flex-direction 전환 | 6차시 |
| `#hero` | flex 중앙 정렬, vh 단위 | 3, 6차시 |
| `#features` | display:grid, auto-fit, minmax | 7차시 |
| `#about` | grid 2열 레이아웃 | 7차시 |
| `#gallery` | flex-wrap | 6차시 |
| `#contact` | 폼 시맨틱 태그, CSS 스타일링 | 2, 3, 5차시 |
| `<footer>` | flex-wrap, column→row | 6차시 |
| 전체 | CSS 변수, 미디어 쿼리 | 3, 8차시 |
| 전체 | 시맨틱 HTML | 1차시 |

## 9. 마무리 체크리스트

### HTML 구조 점검

- [ ] 시맨틱 태그를 적절히 사용했는가? (`header`, `nav`, `main`, `section`, `footer`)
- [ ] `<meta charset="UTF-8">` 선언이 있는가?
- [ ] `<meta name="viewport" ...>` 선언이 있는가?
- [ ] 이미지에 `alt` 속성을 모두 작성했는가?
- [ ] 폼 요소에 `label`이 연결되어 있는가?

### CSS 스타일 점검

- [ ] CSS 선택자 충돌(specificity 문제)이 없는가?
- [ ] 모바일 기본 스타일이 먼저 작성되어 있는가? (모바일 퍼스트)
- [ ] `max-width` + `margin: 0 auto`로 컨테이너를 가운데 정렬했는가?
- [ ] 이미지에 `max-width: 100%`가 적용되어 있는가?
- [ ] CSS 변수(Custom Properties)로 색상을 일관성 있게 관리했는가?

### 반응형 동작 점검

- [ ] 576px, 768px, 1024px, 1200px 기준으로 레이아웃이 변경되는가?
- [ ] 가로 스크롤이 발생하지 않는가?
- [ ] 텍스트가 너무 작아지거나 겹치지 않는가?
- [ ] 버튼/링크의 터치 영역이 최소 44×44px 이상인가? (모바일 접근성)
