```python
# 제품마다 발생하는 불량코드가 있음
# ERR-01 : 도색불량
# ERR-02 : 표면불량
# ERR-03 : 색상불량
# (ERR-01), (ERR-01, ERR-03), (ERR-01, ERR-02, ERR-03)

# Phase3 가변 매개변수는 보통 가장 맨 뒤에 위치 그러나 키워드 매개변수가 있으면 그 앞에 위치.
def inspect_part_v3(part_id, part_size, *errors,min_size=148.0, max_size=152.0 ):
    if part_size >= min_size and part_size <= max_size:
        print(f'{part_id} : 정상 (크기 : {part_size}cm)')
    else:
        print(f'{part_id} : 불량 (크기 : {part_size}cm)')
        print(f'크기불량 외 추가 불량 정보')
        for err in errors:
            print(f'error : {err}')

inspect_part_v3("Bumpt_BBB", 120)
inspect_part_v3("Bumpt_BBB", 120, 'ERROR-1', 'ERROR-2')
inspect_part_v3("Bumpt_BBB", 120, min_size=120, max_size=145)
inspect_part_v3("Bumpt_BBB", 120, 'ERROR-1', 'ERROR-2', min_size=120, max_size=145)
```

```python
# 제품마다 발생하는 불량코드가 있음
# ERR-01 : 도색불량
# ERR-02 : 표면불량
# ERR-03 : 색상불량
# (ERR-01), (ERR-01, ERR-03), (ERR-01, ERR-02, ERR-03)

# Phase3 가변 매개변수는 보통 가장 맨 뒤에 위치 그러나 키워드 매개변수가 있으면 그 앞에 위치.
def inspect_part_v3(part_id, part_size, *errors, min_size=148.0, max_size=152.0 ):
    if part_size >= min_size and part_size <= max_size:
        print(f'{part_id} : 정상 (크기 : {part_size}cm)')
    else:
        print(f'{part_id} : 불량 (크기 : {part_size}cm)')

    if errors:
        print(f'===크기불량 외 추가 불량 정보===')
        for err in errors:
            print(f'error : {err}')
```

```python
# 품질팀에서 요구사항 : 검사할 때 공장의 온도, 습도, 작업자 이름, 라인번호 등 환경데이터도 같이 DB에 저장해주세요.
# 이러한 요구사항은 매번 바뀔 수 있음.. 그래서 **env_data 로 처리하기로함
# 리팩토링 환경데이터는 *defects, **env_data
def inspect_part_final(part_id, mesure_size, *defects, **env_data,):
  print(f'{part1_id} 종합 검사 로그 저장')
  # 1. 비정형 환경데이터
  if env_data:
    print('환경 데이터')
    for key,value in env_data.items():
      print(f"  > {key} : {value}")
  if defects:
    print(f" [불량내역] {defects}")
```
