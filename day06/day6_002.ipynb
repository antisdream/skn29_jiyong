```python
with open('my_diary.txt', 'w', encoding='utf-8') as f:
    f.write('2026년 3월 19일\n')
    f.write('========================\n')
    f.write('오늘 날씨: 맑음\n')
    f.write('오늘의 기분: \n')
    f.write('일과:\n')
    f.write('- 아침 9시: 깨어남\n')
    f.write('- 오전 10시: 파일 입출력 강의 수강\n')
    f.write('- 오후 2시: 간식 시간\n')
    f.write('- 오후 3시: 실습\n')
    f.write('내일의 목표: 파일 입출력 완전 정복!\n')

print("일기가 저장되었습니다.")
```

```python
# 1. 저장된 일기 읽기
with open('my_diary.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
# 2. 줄 단위로 읽기
with open('my_diary.txt', 'r', encoding='utf-8') as f:
    for line in f:
        print(line)
```

- 모드는 w

```python
# poem.txt 파일을 읽어서 라인 번호를 넣어서 출력 (또는 저장)
# 요구사항 enumerate(temp,1)

with open('poem.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()

with open('poem.txt', 'w', encoding='utf-8') as f:
    for index, value in enumerate(lines, 1):
        f.write(f'[{index}] {value}')
```
