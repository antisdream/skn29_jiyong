```python
# csv란
# 쉼표로 구분, 엑셀로 열 수 있고, DB에 자주 나타남, 가볍고 호환성이 좋다
# encoding='utf-8'은 엑셀용, encoding='cp949'는 vscode에서 열 때 사용.
# csv파일은 .csv뒤에 .py를 붙여서 파이썬 환경으로 가능.
# pandas는 csv파일을 읽고 쓰는 명령어


# 이름, 나이, 직급, 월급
# 김철수, 35, 과장, 5000
# 이영희, 28, 대리, 3500
# 박민준, 42, 부장, 7000
```

```python
import csv
with open('employes.csv', 'w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    # 해더 행
    writer.writerow( ['이름', '나이', '직급', '월급'] )
    # 데이터 행
    writer.writerow( ['김철수', 35, '과장', 5000])
    writer.writerow( ['이영희', 28, '대리', 5000])
    writer.writerow( ['박민준', 42, '부장', 5000])
```

```python
# csv 파일을 list로 읽기
import csv
with open('employes.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row, type(row))
```

```python
# csv 파일을 dict로 읽기
import csv
with open('employes.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row, type(row))
```

j-son
- 모든 것을 담을 수 있다.
- 생성 할수도 있고, 생성된 것을 가져다 쓸 수도 있다.
- j-son은 설치하지 않아도 내장되어있다.

```python
# j-son 구조

person = {
    "이름": "김철수",
    "나이": 35,
    "직급": "과장",
    "월급": 5000,
    "스킬": ["Python", "JavaScript", "SQL"],
    "주소": {
        "시": "서울",
        "구": "강남구"
    }
}
import json
# 파일로 json 객체(정보)를 저장
with open('person2.json', 'w' ,encoding='utf-8') as f:
    json.dump(person, f, ensure_ascii=False, indent=2)
    #실무에서 쓰이는 json!
```

```python
import json
with open('person2.json', 'r', encoding='utf-8') as f:
    person2 = json.load(f)
```
