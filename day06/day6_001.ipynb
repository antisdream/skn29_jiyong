### path (경로)
절대경로 r("C:\a\b~~.ipynb") == C:/a/b~~.ipynb 
- 위치가 어디에 있든 경로가 처음 상위폴더부터 잡힌다.
- Ex) C:\python-src\day06\day6_00_review.ipynb
- r''은 \의 기능을 죽이는 역할이다, 그래서 \들이 있는 경로에는 r을 묶어 붙인다.
- r'C:\python-src\day06\day6_00_review.ipynb'

상대경로 r("a\b~~.ipynb") == a/b~~.ipynb
- .day06\001.ipynb


경로설정 방법
- ./ : 현재 위치
- ../ : 상위 폴더
- 1. 절대 경로 C:/python-src/day06/day6_001.ipynb
- 절대 경로가 편하지만 배포에 어렵다.
- 2. 상대 경로 ../day06/day6_001.ipynb
- 상대 경로는 과정이 있지만 배포에 용이하다.

밑의 작업은 ipynb말고 py환경에서 사용 가능!

```python
# 현재 디렉터리 확인
import os
import sys

# 현재 작업 디렉토리를 확인하는 기능
print(f'현재 작업디렉터리 : {os.getcwd()}')

# Pyhton 파일의 위치 확인
print(f'파일위치 확인 : {os.path.abspath(__file__)}')

print(f'__file__ : {__file__}')
```

mode란?
- 내가 쓰려는 목적에 맞게 그 파일을 지정해주면 그 mode를 실행해주고,
- 만약 그 mode가 정의가 되지 않았다면 새로운 mode를 생성해준다.
- r + : 쓰기, 읽기 
- w + : 읽기, 쓰기, 기존내용 삭제

```python
# 파일열기
# 기본문법
# (file:FileDescriptor0rPath) : 
filename = r'da06\day6_001.py'
# filename을 입력할 때 이미 day06폴더에 있는것을 알고 있기 때문에 상위폴더는 앞에 안붙여도 괜찮음
mode = 'r'
# mode는 값을 안주면 기본값이 r임
file_obj = open(filename, mode, encoding='utf-8')
# utf-8 대신 cp949가 쓰이기도 함
# parameter은 값을 안줘도 동작을 함
file_obj.close() 
# 파일을 열었으면 다시 닫아줘야 함
```

# With 구문을 사용하면 파일 누수(file leak)

```python
# file을 열고 닫는것은 예전 방식이다.
# 요즘엔 사용하고 자동으로 닫히게 만든다.. --> with 자동 close

#  with 구문을 빠져나오면 자동으로 close...
with open('text.txt', 'r') as f_read:
    content = f_read.read() #로직이 자리를 위치함
    print(content)
```

```python
# 파일에 기록하기 (기존에 있던것을 없애고 덮어씀)
# 기존에 파일이 없으면 파일을 만들고 텍스트를 입력해준다.
with open('text.txt', 'w') as f_write:
    content = '안녕하세요'
    f_write.write(content)
```

```python
# 파일 읽는방법
# 1. 전체 내용을 한번에 가져오기
# 2. 한 줄씩 읽기
# 3. 모든 줄을 리스트로 읽기
# 4. 반복문을 이용해서 줄 단위로 읽기 --> 가장 추천하는 방법
```

```python
# 1 전체 내용을 한번에 가져오기
with open('text.txt', 'r') as f:
    content = f.read()
    print(content)
```

```python
# 2 한 줄씩 읽기(파일안에 txt의 data가 많을 경우에는 컴퓨터가 먹통이 될수도 있으니 한줄씩 가져온다)
with open('text.txt', 'r') as f:
    line = f.readline()
    print(line)
# 순환문 while을 사용해서 공백(글이 끝날때까지)이 나올때까지 순환시킨다
```

```python
# 2-1 (파일안에 txt의 data가 많을 경우에는 컴퓨터가 먹통이 될수도 있으니 한줄씩 가져온다)
with open('text.txt', 'r') as f:
    print(f.readline())
# 순환문 while을 사용해서 공백(글이 끝날때까지)이 나올때까지 순환시킨다
# print(f.readline()) 개수 = 출력하는 줄의 수
```

```python
# 3 (전체를 한 줄에 리스트로 묶어서 가져올 때 씀)
with open('text.txt', 'r') as f:
    lines = f.readlines()
    print(lines)
    print(type(lines))
```

```python
# 4 (반복문을 이용해서 줄 단위로 읽기)
with open('text.txt', 'r') as f:
    for line in f:
        print(line)
```

```python
# 4-1(반복문을 이용해서 줄 단위로 읽기)
with open('text.txt', 'r') as f:
    for line in f:
        print(line.strip())
        # \n을 제거하여 한 줄씩 올라온다
```
