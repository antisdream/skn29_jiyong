```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 전처리로 깔끔한 DB구축이 우선
# 데이터 준비
x = np.array([1, 2, 3, 4]).reshape(-1, 1) # 공부시간
y = np.array([50, 60, 70, 80])
print(x.shape, y.shape)

# 모델생성
model = LinearRegression()

# 모델학습
# fit에게는 학습용 데이터와 정답을 줘야한다
# 머신러닝은 2차원 데이터를 원한ek.
model.fit(x,y)

# 학습결과
print(f'기울기 : {model.coef_}')
print(f'절편 : {model.intercept_}')

new_x = np.array([5, 10, 3.5, 20]).reshape(-1, 1)
print(f'새로운데이터 x에 대한 모델이 예측한 값 : {model.predict(new_x)}')

# y = 10x + 40
# 학습했을때 데이터 모양이 2차원이었기 때문에 학습한 데이터와 똒같은 shape모양으로 물어봐야 한다.
```

# [2교시]

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x, y, color='blue', s=100, label='real_data') 
# 머신러닝이 예측한 예측직선
x_line = np.linspace(0, 6, 100).reshape(-1, 1)
y_line = model.predict(x_line)
ax.plot(x_line, y_line, color = 'red', label = 'predict_data')
# 새로운 예측값
new_x = np.array([[4.5]])
new_y = model.predict(new_x)
ax.scatter(new_x, new_y, color = 'green', label = 'new_data')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# age: 나이
# sex: 성별
# bmi: 체질량지수 (Body Mass Index)
# bp: 평균 혈압 (Average Blood Pressure)
# s1 ~ s6: 6가지 혈청 수치 (혈액 검사 결과)
# s1: tc (Total Serum Cholesterol) - 총 콜레스테롤
# s2: ldl (Low-Density Lipoprotein) - 저밀도 지단백
# s3: hdl (High-Density Lipoprotein) - 고밀도 지단백
# s4: tch (Total Cholesterol / HDL) - 총 콜레스테롤/HDL 비율
# s5: ltg (Log of Triglycerides) - 트리글리세리드 로그 수치
# s6: glu (Blood Sugar) - 혈당 
# 모든 컬럼은 평균 0, 표준편차 1로 정규화(standardized)되어 있음
```

```python
load_diabetes(as_frame=True).frame

# 정규화(standardized) : 각 원소를 평균으로 뺀 값을 표준편차로 나눔

data = np.array([10, 20, 30])
data.mean(), data.std()

standard_data = (data - data.mean()) / data.std()
standard_data.mean(), round(standard_data.std(), 1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

data = tips['total_bill'].values
fig, axes = plt.subplots(1, 2, figsize = (8, 6))
axes[0].hist(data)

data_mean = data.mean()
data_std = data.std()
standard_data = (data-data_mean) / data_std
axes[1].hist(standard_data)
```
