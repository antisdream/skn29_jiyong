## 절대 오차의 평균

$$ MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i) $$


- 이상치에 덜 민감한 지표가 필요할 때, 오차를 직관적으로 파악하기 쉽다

## 예측 오차의 제곱 평균 값

$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

- 오차를 제곱 : 1미만의 오차는 작아지고 1이상의 오차는 매우 커짐
- 모델이 오차가 큰 예측에 대해 큰 패널티를 부여 - 주로 사용
- 이상치를 제거하는게 아니라 이상치가 중요한 도메인 : 금융리스크, 의료진단

```python
0.5**2, 1.5**2, 1.5-0.5, 2.25-0.25
```

## 설명력 (모델이 타겟 분산을 얼마나 설명하는지)

$$RMSE =  \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

- MSE에 제곱을 했으니 원래 데이터의 SCALE이 변함, 그래서 맞춤
- MSE의 장점을 그대로 가져가면서 MAE 장점도 가져감(오차가 큰 가중치에 패널티를 더 크게 하면서오차를 직관적으로 파악)

$$R^2 = 1 -  \frac{SSE}{SST}= 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$$

- 데이터의 분산을 설명하는 상대적인 성능(0 ~ 1)
- SSE(Sum of Squared Errors) : 오차 제곱의 합 ( 잔차의 합 )
- SST(Total Sum of Squares)
    - 실제값들의 평균값($hat{y}$)과 실제값들의 차이 제곱합

- $\hat{y}$ : (10+20+30) / 3 = 20
- 실제 : 10 20 30
- 예측 : 12 18 33
- SST : $(10-20)^2$ = 100
    - 100 + 0 + 100 = 200
- SSe : 17
- 1 - 14/200 = 0.915

- 0 ~ 1 사이의 측정지표     객관적인 성능비교

# [2교시]

- 데이터 분할 및 표준화 코드

```python
from sklearn.model_selection import train_test_split  # 각주: 데이터셋을 학습/테스트로 나누는 함수 임포트
from sklearn.preprocessing import StandardScaler      # 각주: 특성 스케일링(표준화)을 위한 클래스 임포트

# --- (전제) x, y가 올바른 NumPy 배열 또는 pandas 객체여야 합니다.
# 예: X = data.iloc[:, :-1].to_numpy(); y = data['petal width (cm)'].to_numpy()

# 학습/테스트 분할
x_train, x_test, y_train, y_test = train_test_split(
    x,                # 각주: 입력 특징(설명 변수)
    y,                # 각주: 타깃(예측하려는 값)
    test_size=0.25,   # 각주: 테스트 데이터 비율(생략하면 기본 0.25). 필요에 따라 조정 가능.
    random_state=42   # 각주: 난수 시드 고정으로 재현성 보장
)

# 스케일러 인스턴스 생성
scaler = StandardScaler()  # 각주: 평균 0, 표준편차 1로 변환하는 스케일러

# 학습 데이터에 대해 fit(통계 계산)과 transform(변환)을 동시에 수행
x_train = scaler.fit_transform(x_train)
# 각주: fit_transform은 학습 데이터의 평균/분산을 계산한 뒤 그 통계로 변환합니다.
# 각주: 절대 금지 — 테스트 데이터로 fit을 호출하면 정보 누수(data leakage)가 발생합니다.

# 테스트 데이터는 학습 데이터에서 계산한 통계로만 transform
x_test = scaler.transform(x_test)
# 각주: transform만 호출하여 학습 데이터의 평균/분산을 사용해 동일한 스케일링을 적용합니다.
```

- XGBoost 회귀

```python
from xgboost import XGBRegressor            # 각주: XGBoost의 회귀 모델 클래스 임포트

xgb = XGBRegressor()                        # 각주: 기본 하이퍼파라미터로 모델 인스턴스 생성

xgb.fit(x_train, y_train)                   # 각주: 학습 데이터(x_train, y_train)로 모델 학습
                                            # 각주: 필요하면 eval_set, early_stopping_rounds 등을 추가해 검증 모니터링 가능

r2_score(y_test, xgb.predict(x_test))       # 각주: 테스트 데이터에 대한 예측을 수행하고 R²(결정계수)를 계산
                                            # 각주: xgb.predict(x_test)는 NumPy 배열을 반환하므로 sklearn의 r2_score와 바로 호환됨
```

- PyTorch 데이터 준비 및 DataLoader

```python
import torch                                 # 각주: PyTorch 핵심 패키지 임포트
import torch.nn as nn                        # 각주: 신경망 모듈(nn) 임포트
from torch.optim import Adam                 # 각주: Adam 옵티마이저 임포트
from torch.utils.data import TensorDataset, DataLoader  # 각주: Dataset/DataLoader 임포트
from torch.optim import lr_scheduler         # 각주: 학습률 스케줄러 모듈 임포트

torch.manual_seed(42                          # 각주: 재현성을 위한 시드 고정
)
x_train_t = torch.FloatTensor(x_train)        # 각주: NumPy/배열 -> FloatTensor로 변환 (학습 입력)
x_test_t = torch.FloatTensor(x_test)          # 각주: 테스트 입력을 FloatTensor로 변환
y_train_t = torch.FloatTensor(y_train)        # 각주: 학습 타깃을 FloatTensor로 변환
y_test_t = torch.FloatTensor(y_test)          # 각주: 테스트 타깃을 FloatTensor로 변환

train_dataset = TensorDataset(x_train_t, y_train_t)  # 각주: (입력, 타깃) 쌍으로 학습용 Dataset 생성
test_dataset = TensorDataset(x_test_t, y_test_t)    # 각주: 테스트용 Dataset 생성

train_loader = DataLoader(train_dataset, batch_size = 36, shuffle = True)  # 각주: 학습용 DataLoader (배치, 셔플)
test_loader = DataLoader(test_dataset, batch_size=32)                     # 각주: 평가용 DataLoader (셔플 기본 False)
```

- 신경망 모델 정의

```python
class IrisReg(nn.Module):
    def __init__(self, input_dim, *args, **kwargs):
        super(). __init__(*args, **kwargs)            # 각주: 부모 클래스(nn.Module) 초기화. 오타 없이 정확히 호출해야 함.
        self.network = nn.Sequential(
            nn.Linear(input_dim, 256),                # 각주: 입력 차원 -> 256 유닛 선형층
            nn.BatchNorm1d(256),                      # 각주: 배치 정규화(특성별 평균/분산 정규화)
            nn.ReLU(),                                # 각주: 비선형 활성화 함수
            nn.Dropout(0.2),                          # 각주: 드롭아웃(과적합 방지, 확률 0.2)
            nn.Linear(256, 128),                      # 각주: 256 -> 128 선형층
            nn.BatchNorm1d(128),                      # 각주: 128 채널에 대한 배치 정규화
            nn.ReLU(),                                # 각주: ReLU 활성화
            nn.Dropout(0.2),                          # 각주: 드롭아웃(확률 0.2)
            nn.Linear(128, 32),                       # 각주: 128 -> 32 선형층
            nn.BatchNorm1d(32),                       # 각주: 32 채널에 대한 배치 정규화
            nn.ReLU(),                                # 각주: ReLU 활성화
            nn.Dropout(0.5),                          # 각주: 드롭아웃(확률 0.5, 더 강한 정규화)
            nn.Linear(32, 1)                          # 각주: 출력층(회귀: 1차원 실수 출력)
        )
    def forward(self,x):
        return self.network(x)                       # 각주: 입력 x를 네트워크에 통과시켜 출력 반환
```

- Pytorch 회귀 신경망 학습 및 평가 파이프라인

```python
from tqdm import tqdm
from torch.optim import lr_scheduler

model = IrisReg(x_train_t.shape[-1])                 # 각주: 모델 인스턴스 생성 (입력 차원 전달)
optim = Adam(model.parameters(), lr=0.01)            # 각주: Adam 옵티마이저 생성, 학습률 0.01
scheduler = lr_scheduler.PolynomialLR(optim, total_iters=10)  # 각주: 다항 스케줄러 생성(total_iters는 예시값)

epochs = 100                                         # 각주: 전체 에폭 수
criterion = nn.MSELoss()                             # 각주: 회귀용 MSE 손실 함수

model.train()                                        # 각주: 학습 모드로 전환 (BatchNorm/Dropout 동작)
for epoch in tqdm(range(epochs)):
    for batch_x, batch_y in train_loader:
        optim.zero_grad()        
        predict = model(batch_x).squeeze(1)          # 각주: 모델 출력 (shape: [batch,1]) -> squeeze로 [batch]로 변환
                                                      # 각주: 만약 batch가 1일 때 squeeze가 차원 소실을 유발할 수 있으니 주의
        loss = criterion(batch_y, predict)           # 각주: 현재는 (target, prediction) 순서로 호출됨
                                                      # 권장: criterion(prediction, target) 순서로 사용하세요.
        loss.backward()
        optim.step()
        # scheduler.step()                           # 각주: 스케줄러 스텝 호출은 보통 에폭 끝에서 수행합니다.
                                                      # 만약 배치마다 LR을 바꾸려면 여기서 호출할 수 있으나
                                                      # PolynomialLR의 total_iters 설정과 호출 빈도를 맞춰야 합니다.

model.eval()                                         # 각주: 평가 모드로 전환 (BatchNorm/Dropout 고정)
with torch.no_grad():
    test_predict = model(x_test_t)                   # 각주: 테스트 입력으로 예측 (텐서 반환)
    print(r2_score(y_test_t, test_predict)  )        # 각주: sklearn의 r2_score는 NumPy 배열을 기대합니다.
                                                      # 권장: r2_score(y_test_t.cpu().numpy(), test_predict.cpu().numpy())
```
