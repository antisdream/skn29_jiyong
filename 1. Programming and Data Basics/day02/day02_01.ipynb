데이터 타입과 조건문(if) 결합
- if, elif, else
- Truthy, Falsy : bool 타입이 아니여도 조건문에서 참/거짓으로 평가되는 값들이 있음
- 숫자 0, 빈 문자열 " ", 빈리스트 [] 등은 False로 판단

```python
status_code = 200
error_message = ""

if status_code == 200:
    print("데이터 수집 성공")
elif status_code == 400:
    print("페이지를 찾을 수 없음")
else:
    print("알 수 없는 오류")

if not error_message:
    print("에러 로그가 비었습니다. 정상 작동 중.")
```

논리연산
- and : 조건이 하나라도 False이면 전체가 False (~이고 ~이다 싸고 맛있는 집)
- or : 조건이 하나라도 True이면 전체가 True (~ 또는 싸거나 맛있는 집)
- not : 

```python
score = 92
# 점수가 80이상이고 90점 미만
is_valid = (score >= 80) and (score < 90)
print(f"조건 만족 여부 : {is_valid}")
```

```python
# 수집된 배터리 센서 데이터
battery_temp = 75.5
battery_voltage = 11.2
is_charging = True

#1. 정상조건 : 온도가 80도 미만이고, 전압이 12v이상이어야 정상
is_normal = (battery_temp < 80) and (battery_voltage >= 12)
#2. 온도가 90도 이상이거나, 전압이 10v 미만이면 경고 발생
is_warning = (battery_temp >= 90) or (battery_voltage < 10)
#3. 유지보수필요(복합연산) 충전중이 아닌데, 전압이 11.5v 미만이면 점검필요
is_maintenance_needed = (not is_charging) and (battery_voltage < 11.5)

print(f"배터리 상태 - 정상: {is_normal}, 경고: {is_warning}, 유지보수 필요: {is_maintenance_needed}")
```

```python
# 수집된 배터리 센서 데이터
battery_temp = 75.5
battery_voltage = 11.2
is_charging = True

#1. 정상조건 : 온도가 80도 미만이고, 전압이 12v이상이어야 정상
if (battery_temp < 80) and (battery_voltage >= 12):
    print("정상")

#2. 온도가 90도 이상이거나, 전압이 10v 미만이면 경고 발생
elif (battery_temp >= 90) or (battery_voltage < 10):
    print('경고')

#3. 유지보수필요(복합연산) 충전중이 아닌데, 전압이 11.5v 미만이면 점검필요
elif (not is_charging) and (battery_voltage < 11.5):
    print('점검필요')
```

딕셔너리 { }
- key : value
- key는 중복안됨, value는 제한 없음
- 주민등록번호 : 이름

```python
device_config = {
    'device_id' : 'N-4090',
    'location' : 'A-Block',
    'status' : 'Online',
    'sampling_rate' : 0.5
}
student_config = {
    'name' : '홍길동',
    'kor' : 100,
    'eng' : 95,
    'math' : 100,
    'addr' : '서울시 강남구'
}
# Key : Value 
#변수 하나가지고 class하나를 cover한다.
# read
print(f'device id : {device_config["device_id"]}')
# update 학생 이름은 홍길동 -> 길동이
student_config['name'] = '길동이'
print(student_config) 
#delete
print(student_config.pop('name'))
print(student_config)
#create dictionary에 name이 있으면 '철이'로 바뀌고, 없으면 새로 추가된다.
student_config['name'] = '철이'
print(student_config)
student_config.keys() # list로 key값을 출력
student_config.values() # list로 value값을 출력
student_config.items()
```

```python
# student_config['abs'] --> abs라는 값이 없기 때문에 Error
# student_config.get() --> 내가 찾고자 하는 key와 그 key가 없을 때 찾고자하는 default값
student_config.get('name',0)
```
