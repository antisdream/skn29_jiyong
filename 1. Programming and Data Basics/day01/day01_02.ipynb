# 문자열
- 인덱스 슬라이싱

```python
states = "2026-03-12, 30.5, 25, 45%, True"
states_len = len(states)
print(f'문자열 길이 = {states_len}')
# 인덱스  문자열이름[인덱스번호]
print(states[5])
# 슬라이싱  7:3으로 문자열을 분리
split_size = states_len*0.7
print(split_size)
print( int(split_size) )

split_size = int(split_size)
print(states[ : split_size])
print(states[split_size : ])
```
