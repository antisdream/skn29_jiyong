# 클래스
- 변수
- 함수) 즉, 클래스 내부에 정의한 함소는 메소드(method)
- 생성자 : 객체를 생성할 떄 반드시 호출되는 내부 메소드
- class 다음 첫글자는 무조건 대문자로 암묵적으로 지키고 있는 합의
- class 안에 def같은 함수가 들어가면 method라고 함
- 모든 객체는 생성자를 호출 하게 되어있지만, 만들지 않아도 자동으로 내장되어 동작 됨

```python
# 붕어빵 Class
class FishBread:
    # 생성자는 코딩하지 않으면 아래와 같이 자동으로 기본형태가 만들어져서 실행된다
    def __init__(self): # 첫번째()안에는 무조건 self가 들어가야함 ()안에 self뒤에 연속적으로 내용을 적어놓을 수 있다.
        print('생성자 호출')
    def myfunction(self,name=""): #self는 매개 변수 / 함수 안에서 name값을 가져오려면 생성자에 name이 들어가야한다, 그렇지 않으면 name값을 밖에서 넣어준다.
        print("내가 만든 함수", name)
```

```python
fish1 = FishBread()
fish2 = FishBread()
```

```python
fish1.myfunction() # 이것을 쓰려면 name에 default값을 줘야한다(parameter)
```

```python
# 계산기 클래스
# self.으로 시작하는게 인스턴스 변수라고 한다.
class Calculate:
    def __init__(self):
        pass
    def makeInfo(self,x,y):
        self.x = x
        self.y = y
        
    def add(self):
        return self.x + self.y
    def minus(self):
        return self.x - self.y
```

```python
# class함수의 기본
# 객체 생성
# c1이 self 역할을 하는중
c1 = Calculate()
c1.makeInfo(10,20) # 객체의 변수를 생성하는 역할(인스턴스 변수 생성)
print(c1.add(), c1.minus())
```

```python
# 계산기 클래스
# self.으로 시작하는게 인스턴스 변수라고 한다.
class Calculate:
    def __init__(self,x,y): # 값을 주기 싫으면 x와 y에 0을 주면 된다
        self.x = x
        self.y = y
        
    def add(self):
        return self.x + self.y
    def minus(self):
        return self.x - self.y
```

```python
c1 = Calculate(20, 15)

print(c1.add(), c1.minus())
```

```python
# 클래스
# 속성, 메소드
# 속성 : 변수(데이터)
# 메소드 : 행동(액션)

class Employee:
    # 클래스변수(생성자 밖에서 선언)
    # 모든직원은 Google이라는 회사에 다닌다, 즉 똑같다 (회사 이름 공유)
    company_name = 'Google'

    def __init__(self,name,age):
        # 인스턴스 변수(생성자 안에서 선언)
        # 직원마다 이름과 나이가 다르다, 입사할 때 각자 부여받는다
        self.name = name
        self.age = age

    # Employee가 __init__역할을 한다
emp1 = Employee('홍길동', 23) 
emp2 = Employee('이순신', 20)
```

```python
print(emp1.name, emp1.age, emp1.company_name) # 이름변경을 원하면 emp1.name=''
print(emp2.name, emp2.age, emp2.company_name) # company_name이 변수를 뜻하지만 이름을 부여하게되면 인스턴스 변수가 생성이 된다.
```

```python
emp1.company_name = '삼성전자'
```

```python
print(emp1.name, emp1.age, emp1.company_name)
print(emp2.name, emp2.age, emp2.company_name)
```

```python
emp3 = Employee('철이', 45)
print(emp3.name, emp3.age, emp3.company_name)
```
