# 클래스
    - 속성 : 변수
                - 인스턴스 변수
                - 클래스 변수
    - 액션 : 메소드
    - 상속
        - 오버라이딩

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def showInfo(self):
        return f'name : {self.name}  age : {self.age}'
class Student(Person):
    def __init__(self, name, age, kor, eng, math):
        super().__init__(name, age)
        self.kor = kor; self.eng=eng; self.math = math
        self.total = kor+eng+math
        self.avg = self.total / 3
    def showInfo(self):
        return super().showInfo() + f' kor : {self.kor} eng : {self.eng} math : {self.math}'
class Teacher(Person):
    def __init__(self, name, age, *subjects):
        super().__init__(name, age)
        self.subjects = subjects
    def showInfo(self):
        return super().showInfo() + f' subjects : {self.subjects}'
```

```python
s1 = Student('홍길동',age=25,kor=100,eng=95,math=95)
```

```python
student_params = {
    'kor' : 100, 'eng' : 95, 'math' : 98
}
s2 = Student('이순신', 35, **student_params)
s2.showInfo()
```
