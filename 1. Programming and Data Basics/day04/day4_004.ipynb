```python
# 부모 클래스 car
# 속성 : name, speed
# 함수 : drive

class Car:
    def __init__(self,name='',speed=0):
        self.name = name
        self.speed = speed
    def drive(self):
        print(f" {self.name}이 시속 {self.speed}km로 정속 주행합니다.")
```

```python
c = Car('일반',50)
c.drive()
```

```python
# 상속
class SportCar(Car):
    pass
```

```python
s = SportCar()
    s.drive()
```

```python
class Person:
    def __init__(self):
        self.name=''
        print('부모생성자 호출')
    def showInfo(self):
        print(f'name : {self.name}')

class Student(Person):
    def __init__(self):
        super().__init__() #단독이 아닌 상속받는것이기 때문에 이 형식으로 써줘야한다
        print('자식생성자 호출')
```

```python
# 학생 객체 생성
s1 = Student()
print(s1.name)
```

```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        print('부모생성자 호출')
    def showInfo(self):
        print(f'name : {self.name}')

class Student(Person):
    def __init__(self,name,age):
        super().__init__(name,age)
        print('자식생성자 호출')
```

```python
# 학생 객체 생성
s1 = Student('홍길동', 10)
print(s1.name)
print(s1.age)
```

```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def showInfo(self):
        print(f'name : {self.name} age : {self.age}')

#오버라이딩(over write)
class Student(Person):
    def __init__(self,name,age,score):
        super().__init__(name,age) # super().__init__()가 없으면 부모님꺼가 내꺼가 안된다
        self.score = score
    def showInfo(self):
        print(f'name : {self.name} age : {self.age} score : {self.score}')
            # 기존 객체에 점수를 더하고 싶을 때
```

```python
# 학생 객체 생성
s1 = Student('홍길동', age = 10, score = 100)
s1.showInfo()
```
