# 클래스
- 변수와 함수로 이뤄져있다.
- 클래스에서 함수... 즉, 클래스 내부에 정의한 함소는 메소드(method)
- 생성자 : 객체를 생성할 때 반드시 호출되는 내부 메소드
- class 다음 첫글자는 무조건 대문자를 쓰는것으로 암묵적으로 지키고 있는 합의
- class 안에 def같은 함수가 들어가면 method라고 함
- 모든 객체는 생성자를 호출 하게 되어있지만, 만들지 않아도 자동으로 내장되어 동작 됨
- class가 나오면 무조건 그 다음에는 생성자가 들어간다
- (Ex def __init__(self))


```python
# 붕어빵을 가지고 class 예시를 들어보려 한다.
class FishBread:
    # 생성자는 코딩하지 않으면 아래와 같이 자동으로 기본형태가 만들어져서 실행된다
    def __init__(self): # 첫번째()안에는 무조건 self가 들어가야함 ()안에 self뒤에 연속적으로 내용을 적어놓을 수 있다.
        print('붕어빵을 만들어주세요')
    def myfunction(self,name=""): #self는 매개 변수 / 함수 안에서 name값을 가져오려면 생성자에 name이 들어가야한다, 그렇지 않으면 name값을 밖에서 넣어준다.
        print("내가 만든", name)
```

```python
# 위의 class에서 FishBread(붕어빵)을 굽는 과정을 통해 어떻게 되는지 보기 위해서는
# 붕어빵의 객체를 각각 만들어준다.
fish1 = FishBread()
fish2 = FishBread()
```

```python
fish1.myfunction("")
# 입력을 하면 '내가 만든 함수'만 출력이 되는데, name에 값이 없어서 그렇다.
# fish1.myfunction()에서 ()안에 원하는 문자를 적으면 된다.
# 붕어빵을 굽는 과정이기때문에 속재료를 넣으면 좋을듯 하다.
# Ex) "피자", "팥", "크림" 등...
```

```python
fish1.myfunction(name = '피자 붕어빵')
fish2.myfunction(name = '팥 붕어빵')
# 위에 처럼 name에 속재료를 넣어주고 출력을 하면 완성이된다.
```

- Python에서 class를 설계도라 하고 설계도를 이용해 만든 물품을 객체라고 하기도 한다.
- class를 만들 때 무조건 class 다음 나올 단어는 무조건 영어대문자!
- 객체는 class를 가지고 만드는데, 변수에 .을 붙이면 객체가 된다.
- class 선언을 하고 그다음 들어갈 것(생성자)은 객체가 실행될 때 제일 먼저 실행되어 출력이 되는구조이다.
- 하나가 아닌 두개 이상이 될 수가 있기 때문에 객체를 가지고 변수를 만들때는 동일한 class를 받아 a1, a2 또는 b1, b2로 나타낸다.


```python
# 위에 Remind에서 말한 내용에 대한 예시를 보자면...
class Dog:
    def __init__(self, name, bread, age, is_health):
        self.name = name
        self.bread = bread
        self.age = age
        self.is_health = is_health
    def showInfo(self):
        print(f'name:{self.name}, bread:{self.bread}, age:{self.age}, is_health:{self.is_health}')


d1 = Dog('누렁이', '불독', 2, True)
d2 = Dog('뽀삐', '말티즈', 3, True)
d3 = Dog('시베리안', '허스키', 1, False)
#이것들로 객체를 만들 수 있을것이다...
```

```python
# 만들어낸 객체들에 대한 정보를 보고싶으면 위에서 정의한 showInfo기능을 이용하면 된다
d1.showInfo()
d2.showInfo()
d3.showInfo()
# 객체 뒤에 .을 붙이면서 변수가 되어 개의 정보가 출력이 가능해진다
```

```python
# 학원에서 학생들을 관리할 때도 class를 잘 활용할 수 있는데...
# 학원생들의 이름, 나이, 주소, 점수(국어,영어,수학)를 알고 싶어?
# 심지어 점수의 평균과 총점 그리ㅐ고 모든정보를 보고싶다고?
class Student:
    def __init__(self,age,name,addr,kor,eng,math): 
        self.name = name
        self.age = age
        self.addr = addr
        self.kor = kor
        self.eng = eng
        self.math = math

    def getAverage(self):
        return self.getTotal() / 3

    def getTotal(self):
        return self.kor + self.eng + self.math

    def showInfo(self):
        total = self.getTotal()
        average = self.getAverage()
        print(f'name:{self.name}, age:{self.age}, addr:{self.addr}, total:{total}, average:{average:.2f}')




# 이렇게 작성을 하므로써 설계도는 완성이다, 여기에 학생들 정보만 입력해놓으면 언제든 원할 때 꺼내서 학생들 정보를 확인 할 수 있다.
# getTotal로 묶는 이유가 나중에 총점을 구하는 수식이 바뀔 경우 하나만 고쳐주면 된다.
# 함수 순서상 getTotal이 getAverage보다 앞에 나와야하는데 class에서는 상관없다

s1 = Student(18, '홍길동', '서울시 구로구', 98, 97, 85)
s2 = Student(18, '고길동', '서울시 동작구', 97, 98, 84)
s3 = Student(18, '저길동', '서울시 강남구', 99, 96, 95)
# 학생들 신상카드를 여기에 미리 적거나 다음셀에 적어도 상관이 없다.
```

```python
s1.showInfo()
s2.showInfo()
s3.showInfo()
# 이 호출부호를 통해 해당학생의 이름, 나이, 주소, 총점, 평균을 구할 수 있다.
```

```python
# 생성자(def __init__(self))밖에서 만드는 함수는 공용 함수야.
# 쇼핑몰에서 쇼핑카트(장바구니)에 물건을 담아보려고 해.
# 클래스 이름 ShoppingCart로 지어볼게!
# 변수는 items라는 리스트이며, 쇼핑카트주인의 이름이 들어갈거야
# 메소드 add_item

class ShoppingCart:
    def __init__(self,owner):
        self.owner = owner
        self.items = [] #안에 물건을 함수 밖에서 가져왔기 때문에 생성자 ()에는 안들어가!
    def add_item(self,item): # class안에 있기 때문에 첫번째엔 무조건 self로...
        #pass 추가할 물건이 생각 안나면 pass로 넘겨도 좋아
        self.items.append(item) 
        # 장바구니에 담고 나중에 뒤늦게 담고 싶으면 list에서 추가하는 함수인 .append()를 사용!

# 자 이제부터 스토리텔링으로 진행해볼까?
# 철수와 영희가 각각 빈 장바구니를 하나씩 집어 들었다
cs = ShoppingCart("철수")
yh = ShoppingCart("영희")
# 철수와 영희를 각각 이니셜인 cs와 yh로 칭할거야. 그리고 쇼핑카트의 주인인 철수와 영희가 각각 self자리에 들어가면되겠지!

# 철수가 쇼핑몰에서 노트북과 마우스를 장바구니에 담았어.
cs.add_item('노트북')
cs.add_item('마우스')

# 철수의 쇼핑목록을 확인하려면
print(f'{cs.owner}의 장바구니 : {cs.items}')
# 이것을 호출하면 돼, 아마 노트북과 마우스가 담겨있는게 확인이 가능할거야.

# 영희의 쇼핑목록을 확인하려면
print(f'{yh.owner}의 장바구니 : {yh.items}')
# 이것을 호출하면 돼, 영희는 아직 쇼핑몰에서 장바구니에 담지 않았으니 비어있는것을 확인이 가능할거야!
```

슬라임 키우기
- 그다음은 우리가 슬라임을 키우는 게임을 해볼거야!

```python
class Slime:
# class의 이름은 Slime으로 지어보자.
    total_climes = 0
    # 이것은 함수밖에 있는 것으로, 기본값은 0으로 할게!
    def __init__(self, name, color, hp, size):
        self.name = name; self.color = color; self.hp = hp; self.size = size
        Slime.total_climes += 1
    # 여기에는 슬라임에 대한 정보가 들어가, 그리고 게임이 일회성이 아닌 지속성으로 적립이 되는 형식인 +=연산자를 사용하면 돼!    
    def eat(self,food):
        self.hp +=10
        self.size +=5
        print(f'냠냠 {self.name}이 {food}를 먹고 덩치가 커졌어요!(크기 : {self.size})')
    # 여기에는 슬라임이 음식을 먹을때마다 변하는 것에 대한 조건을 넣었어.
    def play(self):
        if self.hp > 10:
            self.hp -= 15
        else:
            print('밥주세요!')
    # 여기에는 슬라임이 놀때마다 변하는 것에 대한 조건을 넣었어!
    def getStatus(self):
        print(self.name, self.color, self.hp, self.size)
    # 여기에는 슬라임의 현재 상태를 알 수 있는 정보를 넣었어.

s1 = Slime('하얀슬라임', '하양', 10, 5)
# 이것은 첫번째로 등록된 슬라임의 기본정보를 저장한거야.

s1.getStatus()
# 이 변수를 통해 s1에 대한 정보를 알 수 있어.

s1.play()
# 이 변수를 통해 슬라임이 play에 대한 조건을 행동해.

s1.eat()
# 이 변수를 통해 슬라임이 eat에 대한 조건을 행동해.
```

ATM기기에서 통장 사용
- 여기서는 통장이 ATM기기에 들어가서 전개되는 것을 보여줄거야.

```python
class BankAccount:
# class 이름을 통장이라 지을게.    
    def __init__(self,accountNumber,bankName,psw,balance):
        self.accountNumber = accountNumber
        self.bankName = bankName
        self.psw = psw
        self.balance = balance
    # 이거는 통장에 대한 기본정보인데, 계좌번호, 은행명, 비밀번호, 잔액이야.
    def deposit(self,amount):
        self.balance += amount
        print(f'{amount}원 입금 완료, 잔액 : {self.balance}')
    # 이거는 통장에서 이체했을때에 대한 조건이야.
    def withdraw(self,amount,psw):
    # 이것은 잔액(돈, 비밀번호)이야.
        if self.balance > amount and self.psw == psw:
            self.balance -= amount
            print(f'{amount}원 출금 완료, 잔액 : {self.balance}')
        # 여기 if에서 우린 조건을 넣었는데, 입금액이 잔액보다 적고 비밀번호가 일치해야 출금을 할 수 있게 설정했어.
        # and 구문은 둘 다 무조건 True가 되어야 행동을 가능하게 만들어줘.    
        elif self.psw != psw: 
            print('비밀번호 오류')
        # 여기 !=은 같지 않다는 뜻인데, 비밀번호를 잘못 입력하면 비밀번호 오류라고 출력이 되는것을 알 수 있을거야.    
        else:
            print('잔액 부족')
        # 이것은 출금하려는 돈보다 잔액이 더 적을때 잔액이 부족하다는 오류가 출력이 되는것을 알 수 있을거야.

ba = BankAccount("123-123", 'SK', '1234', 10000)
# ba라는 은행계좌를 만들거야, 계좌번호와 은행명 비밀번호 잔액이 들어가겠지?
```

```python
ba.deposit(50000)
# 통장을 개설하고나서 ba라는 은행계좌에 50000을 보낼거야. 그러면 입금완료와 함께 누적된 잔액을 확인가능할거야!
ba.withdraw(60000, '1235')
# 비밀번호를 모르고 마지막글자 잘못 입력해버렸네? 그러면 비밀번호 오류라서 출금이 안돼.
ba.withdraw(50000, '1234')
# 지폐가 필요해서 입금한 돈을 다시 출금하게 되었어. 잔액이 출금액보다 많고 비밀번호가 일치해서 출금이 될거야!
```

세탁기 코스 설정
- 레버를 돌려 코스1~5까지 설정할 때

```python
class WashMachine:
# 지금까지 잘 따라왔어, class명을 WashMachine으로 지을거야.
    def __init__(self):
        self.course = 1
    # 세탁코스가 1부터 5까지 있으니 우린 기본값을 1로 설정했어!
    def turnRight(self): 
        if 1<= self.course < 5:
            self.course += 1
        # 여기서 if 조건은 코스가 1과 같거나 크고 5보다 작는걸 넣었어.
        # 그리고 코스가 1다음은 2가 되어야하니, +=연산자를 사용한거야.
        else:
            self.course = 1
        print(f'세탁코스 : {self.course}')
        # 기본값이 1이라고 했지? 이제 우리가 레버를 돌릴때마다 세탁코스가 2,3,4,5로 변경될거야.
    def turnLeft(self):
        if 1< self.course <=5:
            self.course -= -1
        # 여기서는 우리가 왼쪽으로 돌리면 코스가 1씩 감소가 될거야. 왜냐하면 오른쪽으로 돌렸을 때 1씩 증가하기 때문에 다시 뒤로 돌아가는것도 있어야곘지?
        else:
            self.course = 5
        # 여기도 중요해, 우리가 레버를 돌리다가 1까지 돌렸는데 또 왼쪽으로 돌리면 0이아니라 5가 되어야겠지? 그래서 5라는 조건을 넣은거야.
        print(f'세탁코스 : {self.course}')
```

```python
w1 = WashMachine()
# 세탁기 이름을 설정하고
for i in range(10):
    w1.turnRight()
# 세탁기 레버를 오른쪽으로 10번돌려볼거야!
```

번호대기표
- class에서 어떻게 쓰이는지 확인해볼거야!

```python
class TicketMachine:
# 똑같이 번호대기표 이름을 설정하고
    stand_number = 0
# 전원을 키면 아직 아무도 안뽑았으니 기본값은 0이겠지?
    def pull_tikcet(self):
        TicketMachine.stand_number += 1
    # 번호표를 뽑게되면 1씩 대기번호가 늘어나게되므로 +=연산자를 사용했어.
        print(f'현재대기자는 : {TicketMachine.stand_number}')
    def reset(self):
        TicketMachine.stand_number = 0
    # 기계의 전원을 끄거나 초기화를 하게되면 다시 0으로 돌아가게 설정한거야.
```

```python
a = TicketMachine()
a.pull_tikcet()
# a라는 번호대기표에서 1명이 처음으로 번호표를 뽑았으므로 현재 대기자는 1명이야.
```

```python
a.reset()
# 초기화를 하게되면 대기번호는 다시 0번 기본값으로 돌아가.
```

```python
for i in range(10):
    a.pull_tikcet()
# 번호대기표도 정상적으로 작동되는걸 확인했겠다.. 이제 영업이 시작되어 손님이 들이닥치기 시작했어.
# 순식간에 10명이 번호대기표를 뽑을거야.
```
