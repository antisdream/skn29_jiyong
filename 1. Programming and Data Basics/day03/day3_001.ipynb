````python
import time # 기본 Python이 가지고 있지 않은 기능(라이브러리)을 추가
battery_temp = 35.0
safe_limit = 40.0

print('배터리 온도 모니터링을 시작합니다.')

while battery_temp < safe_limit:
    print(f'[정상] 현재 온도 : {battery_temp: .1f} C')
    # 실제상황에서는 값을 읽어오지만 여기서 임의로 setting
    battery_temp += 0.5
    time.sleep(1) # 1초씩 지연을 시키는 기능

print(f'[경고] 온도 한계치 도달 현재온도 : {battery_temp}c')
````

````python
# 3. 매개변수만 있는 경우

def say_hello(name, age):
    is_adult = age >= 18
    print(f'{name}씨 안녕하세요 당신의 나이는 {age}이군요 성인여부 : {is_adult}')

say_hello('홍길동', 35)
````

````python
# 연비 계산기
def calculate_efficency(distance, fuel_used):
    # 주행거리를 / 사용한 연료량으로 나눔
    eff = distance / fuel_used
    return f'연비 : {eff: .3f} km/L'
````

````python
print( calculate_efficency(300, 20) )
````

````python
print( calculate_efficency(20, 300) )
````

````python
# 함수를 쓰면 좋은점 : 코드를 재활용 할 수 있다.

def choice_menu():
    print('메뉴를 선택하세요')
    print("1. 입력 3. 조회")
    print('2. 출력 4. 종료')
    choice = int(input('번호를 선택하세요'))
    return choice

while True:
    choice = choice_menu()
    if choice == 1:
        print('정보를 입력하세요')
    elif choice == 2:
        print(" 사용자 정보 시스템 입니다.")
    elif choice == 3:
        print(" 조회할 번호를 입력하세요")
    elif choice == 4:
        print(" 프로그램을 종료합니다.")
        break
````

# 키워드 인자(Keyword Arguments)
- 함수를 호출 할 때 순서에 의존하지 않아도 된다.

````python
# 머신러닝 학습
def train_model(epochs, learning_rate, batch_size):
    return f'학습시작 : [에포크{epochs}], 학습률 : {learning_rate} 배치 : {batch_size}]'
````

````python
print(train_model(10, 0.5, 256))
````

````python
print(train_model(learning_rate=0.5, batch_size=30, epochs=5))
````

````python
print(train_model(10, batch_size=30, learning_rate=0.5))
````

````python
# N개 센서의 오차율 평균 구하기
# 변수에 *가 붙으면 집합이라는 뜻
def get_average_error(*erros):
    print(type(erros), erros)

get_average_error(1,2,3)
    
````

````python
def get_average_error(*erros):
    if not erros:
        return 0
    total = sum(erros)
    return total / len(erros)

get_average_error(1,2,3,4,5)
    
````
