```python
# 네비게이션 
def set_navi(destination, route='최적길'):
    print(f'{destination}으로 안내를 시작합니다. (경로:{route})')

# 목적지만 말할 때 (아무말 안하면 최적길!)
set_navi('강릉')
# 특별히 다른길을 원할 때 (무료도로)
set_navi('강릉','무료도로')
# 강릉을 가기위한 네비게이션 호출
set_navi(route='무료도로',destination='강릉')
```

```python
# 휴게소 간식 먹기(for문 결합)
# 간식박스 snacks --> 몇개를 담을지 모른다
def buy_snacks(*snacks):
  print(f'총 {len(snacks)}개의 간식을 샀습니다.')
  # 샀으니 먹어야... for문으로 하나씩 꺼내 먹는다
  for item in snacks:
    print(f'냠냠, {item} 맜있게 먹는다')
  print("배불러요.~~ ^^")
```

```python
# 주행로그를 처리하는 함수
def save_driving_log(vehicle_id, **log_data):
  print(f'-------- 차량 {vehicle_id} 주행 로그 --------- ')
  for key, value in log_data.items():
    print(f' {key} : {value}')
```

```python
save_driving_log('CAR-01', weather='비', temparature=22, speed=120)
```

```python
def func1(model, *trains):
    print(f'학습할 모델명 : {model}')
    print(f'학습에 사용하는 데이터의 개수는 : {len(trains)}') #trains : 튜플
    for i in trains:
        print(f'개별데이터 : {i}')

func1('Tree', 10,20,30,40)
```

```python
def func2(model, **hyper_params):
    print(f'사용한 모델 : {model}')
    print(f'모델의 성능을 향상시키기 위한 하이퍼 파라메터는')
    for key, value in hyper_params.items():
        print(f'{key}:{value}')

func2('Tree', epochs=100, lr = 0.02, batch_size=256)
```
