```python
# 순환문 과 함수
# 매개변수 : 기본이 positional 즉 순서가 중요
# 사용자가 순서를 원하는대로 입력하려면.. 반드시 key:value 형태로 

# 사용자가 값을 주지 않으면 기본값으로 대처하는 기본 매개변수

# 사용자가 몇개의 값을 줄지 모르는 상황이면 하나의 변수로 다 받아들이는 가변매개변수 *변수명
```

```python
# 자동차 조립
def make_car(body_type, wheel_count): # SUV,세단 / 타이어 몇개
    return f"완성! 차체 : {body_type} 바퀴 : {wheel_count}"
```
