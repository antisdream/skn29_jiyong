```python
# 함수 이름 dispatch_robotaxi
# 자율주행 로보택시
# 관제센터로 들어오는 운행기록(log) 제각각, 혼자타는 사람, 5명 탑승, 비 오는날, 배터리가 없는날, 센서가 고장난 날
# 모든 변화무쌍한 변수를 하나의 함수로 통제하는 배차시스템의 함수
# 1. 위치 : 차량번호, 목적지(배차에 절대 빠지면 안되는 필수 정보)
# 2. 디폴트 : 운행모드(설정안하면 컴포트모드로 운행)
# 3. 가변(*args) : 탑승객 명단(1명, 2명, 4명, 빈차일수도 있음)
# 4. 키워드 가변(**kwargs) : 주행환경 및 센서데이터(어떤 데이터가 추가될지 예측불허)
```

```python
def dispatch_robotaxi(id, destination, *args, mode='컴포트', **env_info):
    print(f'기본정보 차량id : {id} 목적지 : {destination}')
    print('탑승객 명단')
    for person in args:
        print(person, end=' ')
    print(f'\n주행모드 : {mode}')
    print('주행환경(에러포함)')
    for key, value in env_info.items():
        print(f'{key}:{v~')
```

```python
# 평일
# dispatch_robotaxi('TAXI-001', '강남역', '컴포트모드', '홍길동')

dispatch_robotaxi('TAXI-001', '강남역', '홍길동', '이순신', mode='컴포트')
```
