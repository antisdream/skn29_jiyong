# [4교시]

# try ~ except ~ else ~ final

```python
try:            # 예외가 발생할 수도 있는 코드
    pass 
    # 자원 할당 -- 성공
    # 자원 이용 -- 실패 즉 오류
except:         # 예외 발생 시 처리
    pass
else:           # 예외가 없을 때 실행
    pass 
finally:        # 무조건 실행
    pass 
    # 자원 해제 -- 성공 실패 여부 상관없이 무조건 실행되기 때문에 finally에 속해있어야 한다.

# 1. try에서 에러가 발생하지 않으면 except를 넘기고 else와 finally를 실행한다.
# 2. try에서 에러가 발생하면 except가 실행되어 else를 넘기고 finally가 실행된다.
# 3. finally는 에러가 있든 없든 무조건 실행되는 녀석이다.
```

```python
def calculate_average(scores):
    if type(scores) != list:
        raise TypeError("매개변수는 반드시 리스트 형태가 되어야 합니다.")
    if len(scores) == 0:
        raise ValueError("리스트는 비어있으면 안됩니다.")

    return sum(scores) / len(scores)


datas = []

try:
    print(calculate_average(datas))
except Exception as e:
    print(f"error : {e}")

# 매개변수가 무엇이 잘못되었는지, 그 형태에 뭐가 잘못되었는지 원인을 알려준다.    
```

```python
def calculate_average(scores):
    if type(scores) != list:
        raise TypeError("매개변수는 반드시 리스트 형태가 되어야 합니다.")
    if len(scores) == 0:
        raise ValueError("리스트는 비어있으면 안됩니다.")

    return sum(scores) / len(scores)


datas = [100, 'a']

try:
    print(calculate_average(datas))
except (TypeError, ValueError) as e:
    print(f"데이터 처리 오류 : {e}")
except Exception as e:
    print(f"error : {e}")

# 이 경우에는 우리가 에러를 발생시킨것이 아닌, 이 자체에서 에러를 발생시켰다.
```

```python
class MissingDataError(Exception):
    # 필수데이터가 누락되었을 때 발생하는 에러
    def __init__(self, column_name):
        super().__init__(f'치명적인 오류 : {column_name} 컬럼의 데이터가 비었습니다.')

# 데이터
test_data = {'name' : '홍길동', 'age' : "", 'addr' : "seoul"}
for key, value in test_data.items():
    try:
        if value == "":
            raise MissingDataError(key)
        else:
            print(value)
    except Exception as e:
        print(e)
```

```python
#4 5회 로그인 실패시 30초간 로그인 차단

import time

def checkIdPassword(id, psw):
    if id != 'USER' and psw != '1234!@':
        raise ValueError('아이디와 패스워드가 모두 틀렸습니다.')
    if id != 'USER':
        raise ValueError('아이디가 틀렸습니다.')
    if psw != '1234!@':
        raise ValueError('패스워드가 틀렸습니다.')
    return True

max_attempts = 5

while True:
    attempts = 0
    logged_in = False

    while attempts < max_attempts:
        try:
            id = input('아이디: ')
            psw = input('패스워드: ')
            checkIdPassword(id, psw)
        except ValueError as e:
            attempts += 1
            remaining = max_attempts - attempts
            print(f'error: {e} (남은 시도 {remaining}회)')
        else:
            print('로그인 성공')
            logged_in = True
            break

    if logged_in:
        break

    print(f'{max_attempts}회 실패로 30초 동안 로그인 차단합니다.')
    time.sleep(30)
    print('차단이 해제되었습니다. 다시 시도하세요.')
```

```python
#5 5회실패시 30초간 차단 및 재로그인 가능시간 알림

import time

def checkIdPassword(id, psw):
    if id != 'USER' and psw != '1234!@':
        raise ValueError('아이디와 패스워드가 모두 틀렸습니다.')
    if id != 'USER':
        raise ValueError('아이디가 틀렸습니다.')
    if psw != '1234!@':
        raise ValueError('패스워드가 틀렸습니다.')
    return True

max_attempts = 5
block_seconds = 30

while True:
    attempts = 0
    logged_in = False

    while attempts < max_attempts:
        try:
            id = input('아이디: ')
            psw = input('패스워드: ')
            checkIdPassword(id, psw)
        except ValueError as e:
            attempts += 1
            remaining = max_attempts - attempts
            print(f'error: {e} (남은 시도 {remaining}회)')
        else:
            print('로그인 성공')
            logged_in = True
            break

    if logged_in:
        break

    print(f'{max_attempts}회 실패로 {block_seconds}초 동안 로그인 차단합니다.')

    # 초 단위 카운트다운 출력
    for sec in range(block_seconds, 0, -1):
        print(f'차단 해제까지 남은 시간: {sec}초', end='\r')
        time.sleep(1)
    print('\n차단이 해제되었습니다. 다시 시도하세요.')
```

```python
#조회, 적립, 종료 메뉴를 추가하여 관리

customer_db = {
    "C001": {"name": "김철수", "points": 1500},
    "C002": {"name": "이영희", "points": 3200},
    "C003": {"name": "박민수", "points": 0}
}

def 조회(id):
    """주어진 id의 고객 포인트와 이름을 출력"""
    info = customer_db.get(id)
    if info is None:
        print('오류: id가 틀렸습니다.')
        return
    print(f'[{id}] 고객명: {info["name"]}  총 포인트: {info["points"]}')

def 적립(id):
    """기존 적립 로직을 함수로 분리"""
    if id not in customer_db:
        print('오류: id가 틀렸습니다.')
        return
    customer_info = customer_db[id]
    while True:
        add_point = input(f'{id} 고객의 추가 할 포인트를 입력하세요: ').strip()
        if add_point == '':
            print('오류: 숫자를 입력하세요')
            continue
        normalized = add_point.lstrip('-+')
        if normalized.count('.') == 1 and normalized.replace('.', '', 1).isdigit():
            print('오류: 숫자만 입력하세요')
            continue
        if not normalized.isdigit():
            print('오류: 숫자만 입력하세요')
            continue
        try:
            pts = int(add_point)
        except ValueError:
            print('오류: 숫자만 입력하세요')
            continue
        if pts <= 0:
            print('오류: 1포인트 이상 입력하세요')
            continue
        customer_info['points'] += pts
        print(f'--> 적립완료! 총 포인트 : {customer_db[id]["points"]}')
        break

while True:
    print('\n메뉴: 1.조회  2.적립  3.종료')
    choice = input('선택하세요: ').strip()
    if choice == '1':
        id = input('조회할 고객아이디를 입력하세요: ').strip()
        조회(id)
    elif choice == '2':
        id = input('적립할 고객아이디를 입력하세요: ').strip()
        적립(id)
    elif choice == '3':
        print('프로그램을 종료합니다.')
        break
    else:
        print('올바른 메뉴를 선택하세요.')
```

```python
# 강사님과 함께 한 것
# 고객 포인트 적립 시스템
customer_db = {
    "C001": {"name": "김철수", "points": 1500},
    "C002": {"name": "이영희", "points": 3200},
    "C003": {"name": "박민수", "points": 0}
}

# 함수는 최소한만 작동할 수 있게 해줘야 한다
def customerSearch():
    id = input('고객아이디')
    return id, customer_db[id]    

def customeraddPoints(customer_info):
    point = int(input('적립할 포인트'))
    if 0< point < 100:
        customer_info['points'] += point
    else:
        raise ValueError('포인트적립범위는 1 ~ 99 ')
    
def customerInfo(id):
    print(f'{customer_db[id]}')
```

```python
# 강사님과 함께 한 것
# 고객 포인트 적립 시스템
try:
    # [1]직원이 고객 아이디를 조회    
    id, customer_info = customerSearch()

    # [2]추가할 포인트값을 받아서 처리
    customeraddPoints(customer_info)
    customerInfo(id)
    
    # [3] 고객의 정보를 적립하며 출력하는 부분
except KeyError as e:
    print(f'고객아이디를 확인하세요 : {type(e)} {e}')
except ValueError as e:
    print(f'값을 확인하세요 정수여부를 확인 : {type(e)} {e}')
except Exception as e:
    print(f'알수없는 에러 관리자 문의 : {type(e)} {e}') # {type(e)} {e}를 하면 무슨 에러인지 나온다. 
```

class변수와 instance변수의 중요성
- 차이점을 잘 알아야함

위의 고객포인트 적립에 관하여 class로 정의 할 때

```python
# 고객 포인트 적립 시스템 [포인트가 같은 매장 다른지점들과 연동되어야하기 때문에 class변수를 쓴다]

class ManagePoint:
    customer_db = {
        "C001": {"name": "김철수", "points": 1500},
        "C002": {"name": "이영희", "points": 3200},
        "C003": {"name": "박민수", "points": 0}
    }
    def __init__(self, id):
        self.id = id

# 함수는 최소한만 작동할 수 있게 해줘야 한다
    def customerSearch(self):
        return customer_db[self.id]    

    def customeraddPoints(self):
        point = int(input('적립할 포인트'))
        if 0 < point < 100:
            ManagePoint.customer_db[self.id]['points'] += point
        else:
            raise ValueError('포인트적립범위는 1 ~ 99 ')
        
    def customerInfo(self):
        print(f'{ManagePoint.customer_db[self.id]}')

# 객체생성
try:
    mnp = ManagePoint(input('고객아이디'))
    mnp.customeraddPoints()
    mnp.customerInfo()
except:
    pass
```
