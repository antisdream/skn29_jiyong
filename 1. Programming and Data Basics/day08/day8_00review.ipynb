DDL의 핵심 명령어
- 데이터베이스, 테이블 생성 및 삭제
- 테이블 구조 변경(컬럼 추가, 수정, 삭제)
- 제약조건 설정
- CREATE	: 데이터베이스 / 테이블 생성
- ALTER	: 테이블 구조 변경
- DROP	: 데이터베이스 / 테이블 완전 삭제
- TRUNCATE : 테이블 내 모든 데이터 삭제(구조유지)

- 데이터 타입
- 숫자 : int, bigint, decimal(M, D) M: 전체 자리 수, D: 소수점 자리 수, FLOAT DOUBLE: 부동 소수점
- 문자 : VARCHAR(N): N의 최대길이 ==> 가변길이 문자열
- 날짜 / 시간 : DATE(날짜), DATETIME(날짜+시간), TIMESTAMP(자동갱신기능)
- 논리형(bool) : 참 거짓(1/0)

```python
-- 완전한 학생 테이블
-- 학번 (기본키, 자동증가)
-- 이름 (필수)
-- 학과 (필수)
-- 학년 (1~4, 기본값 1)
-- 입학년도 (필수)
-- 이메일 (중복 불가)
-- 전화번호
-- 주소
-- 상태 (재학/휴학/졸업, 기본값 재학)
-- 등록일시 (자동)
-- 수정일시 (자동)
drop database if exists studentmng;
create database studentmng;
use studentmng;
create table students(
	student_id int auto_increment primary key comment '학번',
	name varchar(50) not null comment '학생 이름',
    department varchar(50) not null comment '학과',
	admission_year year not null comment '입학년도',
    -- 학년 1~4 허용, 기본값은 1
    grade tinyint check(grade between 1 and 4) default 1 comment '학년',
    email varchar(100) unique comment '이메일(중복불가)',
    phone varchar(10) comment '전화번호',
    address varchar(200) comment '주소',
    status varchar(20) default '재학' comment '상태',
    created_at timestamp default current_timestamp comment'등록일시',
    updated_at timestamp default current_timestamp
		on update current_timestamp comment'등록일시'
) comment "학생정보테이블";

-- 테이블 구조 확인
desc students;
describe students;

-- 테이블 생성 확인(comment)
show create table students;
```

```python
-- 완전한 학생 테이블
-- 학번 (기본키, 자동증가)
-- 이름 (필수)
-- 학과 (필수)
-- 학년 (1~4, 기본값 1)
-- 입학년도 (필수)
-- 이메일 (중복 불가)
-- 전화번호
-- 주소
-- 상태 (재학/휴학/졸업, 기본값 재학)
-- 등록일시 (자동)
-- 수정일시 (자동)
drop database if exists studentmng;
create database studentmng;
use studentmng;
create table students(
	student_id int auto_increment primary key comment '학번',
	name varchar(50) not null comment '학생 이름',
    department varchar(50) not null comment '학과',
	admission_year year not null comment '입학년도',
    -- 학년 1~4 허용, 기본값은 1
    grade tinyint default 1 comment '학년',
    email varchar(100) unique comment '이메일(중복불가)',
    phone varchar(10) comment '전화번호',
    address varchar(200) comment '주소',
    status varchar(20) default '재학' comment '상태',
    created_at timestamp default current_timestamp comment'등록일시',
    updated_at timestamp default current_timestamp
		on update current_timestamp comment'등록일시',
        
	-- check 제약조건(학년은 1~4) / constraint : 제약하다
    constraint chk_grade check(grade between 1 and 4),
    constraint chk_status check(status in ('재학', '휴학', '졸업'))
) comment "학생정보테이블";


-- 테이블 구조 확인
desc students;
describe students;

-- 테이블 생성 확인(comment)
show create table students;
```

# [7교시]

delete field
- -> 조건문에서 yes or no로 삭제한 것을 보여줌

varchar와 char의 차이
- char:고정길이 문자열, varchar: 가변길이 문자열
- char:길이가 일정한 데이터에 적합, varchar: 길이가 들쑥날쑥한 데이터에 적합

modify와 change의 차이
- modify 컬럼명은 그대로 타입/제약만 변경
- change modify 기능 그대로 컬럼명도 변경 가능

```python
create database university;

use university;

create table students(
	student_id int auto_increment primary key comment 'pk',
    name varchar(20) not null comment '학생이름',
    age int not null comment '학생나이'
    ) comment '학생테이블';
    
create table professors(
	professors_id int auto_increment primary key comment 'pk',
    name varchar(20) not null comment '교수이름',
    age int not null comment '교수나이'
    ) comment '교수테이블';
    
create table course(
	course_id int auto_increment primary key comment 'pk',
	professors_id int comment '담당교수',
    
	name varchar(20) not null comment '과목이름',
    grade tinyint not null comment '학점',
    constraint chk_grade check(grade between 1 and 3),
	-- 담당교수 외래 키 제약조건 설정
    foreign key(professors_id) references professors(professors_id)
		on delete set null	-- 교수 삭제시 null(과목은유지)
        on update cascade	-- 교수 번호 변경시 자동 업데이트

)comment '과목테이블';
-- 교수 : 과목	1 : N

create table enrollments(
	enrollment_id int auto_increment primary key comment '수강번호',
    -- 학생(외래 키)
    student_id int not null comment '학번',
    -- 과목(외래 키)
    course_id int not null comment '과목',
	-- 수강정보(2026-1, 2026-2)
    -- 성적정보(중간고사, 기말고사, 과제점수)
    grade varchar(2) comment '학점', -- A+, A, B+, B, C+, C, D, F
    foreign key(student_id) references students(student_id)
		on delete cascade	-- 학생 삭제시 수강 기록도 삭제
        on update cascade,	-- 학번 변경시 자동 업데이트
	foreign key(course_id) references course(course_id)
		on delete cascade	-- 과목 삭제시 수강 기록도 삭제
        on update cascade,	-- 과목코드 변경시 자동 업데이트
		-- constraint chk_grade check(grade in ('A+', 'A', 'B+', 'B', 'C+', 'C', 'D', 'F'))
        -- unique 제약조건 (같은 학생이 같은 과목을 중복 수강 불가)
	unique key unique_enrollments (student_id, course_id)
) comment '수강 테이블';

-- 테이블 생성 확인
show tables;

-- 컬럼 수정 ADD Column
-- 학생테이블 성별 컬럼 추가
alter table students add column gender char(1) comment '성별 (M/F)';

-- 컬럼 수정(modify column)
alter table students modify column age tinyint; 

-- 컬럼 이름 변경
alter table students change column age student_age tinyint;

-- modify vs change
-- modify 컬럼명은 그대로 타입/제약만 변경
-- change modify 기능 그대로 + 컬럼명도 변경 가능

-- 컬럼 삭제
alter table students drop column gender;

-- constraint chk_grade check( grade in ('A+', 'A', 'B+', 'B', 'C+', 'C', 'D', 'F' ) )
-- unique 제약조건( 같은 학생이 같은 과목을 중복 수강 불가)
unique key  unique_enrollments (student_id, course_id) comment '수강 테이블'; 
```
