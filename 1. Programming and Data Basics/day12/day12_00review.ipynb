```python
import requests #파이썬에서 HTTP 요청을 보내기 위해 requests 라이브러리를 코드에 가져와 사용하겠다는 선언

# 공공데이터 포털에서 크롤링 할 때 입력할 개인인증키
service_key = 'WEATHER_API_KEY_PLACEHOLDER'
# 공공데이터 포털에서 크롤링할 고유 URL, 보통 상세 페이지에 안내되어 있음
url = 'https://apis.data.go.kr/B553881/newRegistlnfoService_02/getnewRegistlnfoService02'
params = {
    'serviceKey' : service_key, #개인인증키
    'registYy' : '2017', #추출할 정보의 연도
    'registMt' : '09', #추출할 정보의 월
    'vhctyAsortCode' : '1', #추출할 정보의 고유 코드
    'registGrcCode' : '1', #추출할 정보의 고유 코드
    'useFuelCode' : '5', #추출할 정보의 고유 코드
    'cnmCode' : '', #추출할 정보의 고유 코드
    'prposSeNm' : '', #추출할 정보의 고유 코드
}
response = requests.get(url,params=params) #requests라이브러리를 사용해 특정 웹 페이지에 (파라미터)데이터를 포함하여 GET 방식으로 HTTP요청을 보내고 결과를 response(변수)에 저장
print(f'코드 : {response.status_code}') # 웹 서버로부터 받은 응답의 상태 코드(Status Code)를 화면에 출력하는 명령어
print(response.text) #응답받은 문자를 출력
print(type(response.text)) #응답받은 문자의 유형을 출력
```

```python
import requests #파이썬에서 HTTP 요청을 보내기 위해 requests 라이브러리를 코드에 가져와 사용하겠다는 선언
import xml.etree.ElementTree as ET #웹에서 구조화된 XML데이터를 파싱하는 작업
import os # 기본 라이브러리인 os 모듈 불러오기
from dotenv import load_dotenv # 환경 변수에 .env 파일에 있는 변수 추가
load_dotenv() # 환경 변수 불러오기

service_key = 'WEATHER_API_KEY_PLACEHOLDER' # 개인 인증키
url = 'https://apis.data.go.kr/B553881/newRegistlnfoService_02/getnewRegistlnfoService02' # 추출할 정보의 url 
params = {
    'serviceKey' : service_key,
    'registYy' : '2026',
    'registMt' : '02',
    'vhctyAsortCode' : '1',
    'registGrcCode' : '1',
    'useFuelCode' : '5',
    'agrde' : '3'
}
response = requests.get(url,params=params)
print(f'코드 : {response.status_code}')
print(response.text) # 문자열 xml 구조

root = ET.fromstring(response.text)

for child in root.find('header'): #Element 구조
    print(child.tag, child.text)

for child in root.find('body'): #Element 구조
    print(child.tag, child.text)

print(root.findtext('body/dtaCO'))
```
